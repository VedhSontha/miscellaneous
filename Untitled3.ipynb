```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

# Open the multi-band GeoTIFF
path = r"C:\Users\vedhr\CODES\ssl4eo_l_oli_tirs_toa_benchmark\ssl4eo_l_oli_tirs_toa_benchmark\0000000\LC08_045030_20190814\all_bands.tif"

with rasterio.open(path) as src:
    data = src.read()  # Shape: (bands, height, width)
    print(f"Shape: {data.shape}, Bands: {src.count}")
    print(f"CRS: {src.crs}, Resolution: {src.res}")

# Example: create RGB composite (Bands 4,3,2)
rgb = np.stack([data[3], data[2], data[1]], axis=-1)
rgb = np.clip(rgb / np.percentile(rgb, 99), 0, 1)
plt.imshow(rgb)
plt.title("Landsat 8 True Color Composite (30m)")
plt.show()
```
