```python
# CELL 1: Load & COMPREHENSIVE Feature Engineering
import pandas as pd
import numpy as np
import lightgbm as lgb
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(f"Train: {train.shape}, Test: {test.shape}")
print(f"Click rate: {train['clicked'].mean():.3f}")

def comprehensive_features(df, agg_stats=None, is_train=True):
    df = df.copy()
    
    # Fix data types
    df['liked'] = df['liked'].replace({'yes': 1, 'no': 0, 'Yes': 1, 'No': 0})
    df['liked'] = pd.to_numeric(df['liked'], errors='coerce').fillna(0)
    df['commented'] = df['commented'].fillna(0).astype(int)
    df['subscribed_after'] = df['subscribed_after'].fillna(0).astype(int)
    
    # ===== BASIC ENGAGEMENT FEATURES =====
    df['watch_percent'] = (df['watch_time'] / df['video_duration']).fillna(0).clip(0, 5)
    df['engagement_score'] = df['liked'] + df['commented']*2 + df['subscribed_after']*3
    df['over_watched'] = (df['watch_time'] > df['video_duration']).astype(int)
    df['watch_complete'] = np.minimum(df['watch_percent'], 1.0)
    df['partial_watch'] = ((df['watch_percent'] > 0.3) & (df['watch_percent'] < 0.9)).astype(int)
    df['quick_exit'] = (df['watch_percent'] < 0.1).astype(int)
    df['full_watch'] = (df['watch_percent'] >= 0.95).astype(int)
    
    # ===== TIME FEATURES =====
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    df['hour'] = df['timestamp'].dt.hour
    df['day_of_week'] = df['timestamp'].dt.dayofweek
    df['day_of_month'] = df['timestamp'].dt.day
    df['month'] = df['timestamp'].dt.month
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_weekday'] = (df['day_of_week'] < 5).astype(int)
    df['is_prime_time'] = df['hour'].isin([18,19,20,21,22]).astype(int)
    df['is_morning'] = df['hour'].isin([6,7,8,9,10]).astype(int)
    df['is_late_night'] = df['hour'].isin([0,1,2,3,4,5]).astype(int)
    
    # ===== VIDEO LENGTH CATEGORIES =====
    df['video_short'] = (df['video_duration'] < 300).astype(int)
    df['video_medium'] = ((df['video_duration'] >= 300) & (df['video_duration'] < 1200)).astype(int)
    df['video_long'] = ((df['video_duration'] >= 1200) & (df['video_duration'] < 3000)).astype(int)
    df['video_very_long'] = (df['video_duration'] >= 3000).astype(int)
    
    # ===== DEVICE FEATURES =====
    df['is_mobile'] = (df['device'] == 'Mobile').astype(int)
    df['is_desktop'] = (df['device'] == 'Desktop').astype(int)
    df['is_tv'] = (df['device'] == 'TV').astype(int)
    df['is_tablet'] = (df['device'] == 'Tablet').astype(int)
    
    # ===== CATEGORY FEATURES =====
    df['is_gaming'] = (df['category'] == 'Gaming').astype(int)
    df['is_music'] = (df['category'] == 'Music').astype(int)
    df['is_education'] = (df['category'] == 'Education').astype(int)
    
    if is_train:
        agg_stats = {}
        
        # USER AGGREGATIONS (Most Important!)
        agg_stats['user'] = df.groupby('user_id').agg({
            'clicked': ['mean', 'sum', 'count', 'std'],
            'watch_time': ['mean', 'std', 'min', 'max', 'median'],
            'engagement_score': ['mean', 'sum', 'max'],
            'watch_percent': ['mean', 'std', 'median'],
            'liked': ['sum', 'mean'],
            'recommended': 'mean',
            'over_watched': 'sum',
            'full_watch': 'sum'
        }).reset_index()
        agg_stats['user'].columns = ['user_id'] + ['user_' + '_'.join(col).strip('_') for col in agg_stats['user'].columns[1:]]
        
        # VIDEO AGGREGATIONS
        agg_stats['video'] = df.groupby('video_id').agg({
            'clicked': ['mean', 'sum', 'count', 'std'],
            'watch_percent': ['mean', 'std', 'median'],
            'engagement_score': ['mean', 'sum'],
            'liked': ['sum', 'mean'],
            'full_watch': 'sum'
        }).reset_index()
        agg_stats['video'].columns = ['video_id'] + ['video_' + '_'.join(col).strip('_') for col in agg_stats['video'].columns[1:]]
        
        # CATEGORY AGGREGATIONS
        agg_stats['category'] = df.groupby('category').agg({
            'clicked': ['mean', 'std'],
            'watch_time': ['mean', 'median'],
            'engagement_score': 'mean',
            'watch_percent': 'mean'
        }).reset_index()
        agg_stats['category'].columns = ['category'] + ['cat_' + '_'.join(col).strip('_') for col in agg_stats['category'].columns[1:]]
        
        # DEVICE AGGREGATIONS
        agg_stats['device'] = df.groupby('device').agg({
            'clicked': ['mean', 'std'],
            'watch_time': 'mean',
            'engagement_score': 'mean'
        }).reset_index()
        agg_stats['device'].columns = ['device'] + ['dev_' + '_'.join(col).strip('_') for col in agg_stats['device'].columns[1:]]
        
        # TIME OF DAY AGGREGATIONS
        agg_stats['time_of_day'] = df.groupby('watch_time_of_day')['clicked'].mean().reset_index()
        agg_stats['time_of_day'].columns = ['watch_time_of_day', 'time_click_rate']
        
        # COMBO AGGREGATIONS
        agg_stats['cat_dev'] = df.groupby(['category', 'device'])['clicked'].mean().reset_index()
        agg_stats['cat_dev'].columns = ['category', 'device', 'cat_dev_rate']
        
        agg_stats['cat_time'] = df.groupby(['category', 'watch_time_of_day'])['clicked'].mean().reset_index()
        agg_stats['cat_time'].columns = ['category', 'watch_time_of_day', 'cat_time_rate']
        
        agg_stats['dev_time'] = df.groupby(['device', 'watch_time_of_day'])['clicked'].mean().reset_index()
        agg_stats['dev_time'].columns = ['device', 'watch_time_of_day', 'dev_time_rate']
    
    # MERGE ALL AGGREGATIONS
    df = df.merge(agg_stats['user'], on='user_id', how='left')
    df = df.merge(agg_stats['video'], on='video_id', how='left')
    df = df.merge(agg_stats['category'], on='category', how='left')
    df = df.merge(agg_stats['device'], on='device', how='left')
    df = df.merge(agg_stats['time_of_day'], on='watch_time_of_day', how='left')
    df = df.merge(agg_stats['cat_dev'], on=['category', 'device'], how='left')
    df = df.merge(agg_stats['cat_time'], on=['category', 'watch_time_of_day'], how='left')
    df = df.merge(agg_stats['dev_time'], on=['device', 'watch_time_of_day'], how='left')
    
    # FILL MISSING VALUES
    global_click_rate = 0.5 if not is_train else train['clicked'].mean()
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64'] and df[col].isnull().any():
            if 'rate' in col or 'mean' in col.lower():
                df[col] = df[col].fillna(global_click_rate)
            elif 'sum' in col or 'count' in col:
                df[col] = df[col].fillna(0)
            elif 'std' in col:
                df[col] = df[col].fillna(0)
            elif 'median' in col or 'max' in col or 'min' in col:
                df[col] = df[col].fillna(df[col].mean() if df[col].mean() == df[col].mean() else 0)
    
    # ===== ADVANCED INTERACTION FEATURES =====
    df['user_video_affinity'] = df['user_clicked_mean'] * df['video_clicked_mean']
    df['user_cat_affinity'] = df['user_clicked_mean'] * df['cat_clicked_mean']
    df['engagement_vs_user_avg'] = df['engagement_score'] / (df['user_engagement_score_mean'] + 1)
    df['watch_vs_user_avg'] = df['watch_time'] / (df['user_watch_time_mean'] + 1)
    df['watch_vs_video_avg'] = df['watch_percent'] / (df['video_watch_percent_mean'] + 0.1)
    df['watch_vs_cat_avg'] = df['watch_time'] / (df['cat_watch_time_mean'] + 1)
    df['rec_x_user_rate'] = df['recommended'] * df['user_clicked_mean']
    df['rec_x_video_rate'] = df['recommended'] * df['video_clicked_mean']
    df['rec_x_cat_rate'] = df['recommended'] * df['cat_clicked_mean']
    
    # USER BEHAVIOR INDICATORS
    df['is_active_user'] = (df['user_clicked_count'] > df['user_clicked_count'].median()).astype(int)
    df['is_engaged_user'] = (df['user_engagement_score_sum'] > df['user_engagement_score_sum'].median()).astype(int)
    df['is_binge_watcher'] = (df['user_watch_time_mean'] > df['user_watch_time_mean'].quantile(0.75)).astype(int)
    df['is_casual_user'] = (df['user_clicked_count'] < 5).astype(int)
    df['user_consistency'] = 1 - df['user_clicked_std'].fillna(0.5)
    
    # VIDEO POPULARITY INDICATORS
    df['is_popular_video'] = (df['video_clicked_count'] > df['video_clicked_count'].median()).astype(int)
    df['is_viral_video'] = (df['video_clicked_sum'] > df['video_clicked_sum'].quantile(0.9)).astype(int)
    df['is_niche_video'] = (df['video_clicked_count'] < 10).astype(int)
    df['video_quality_score'] = df['video_liked_sum'] / (df['video_clicked_count'] + 1)
    df['video_engagement_rate'] = df['video_engagement_score_mean'] / (df['video_clicked_mean'] + 0.1)
    
    # MATCH SCORES
    df['perfect_match'] = ((df['user_clicked_mean'] > 0.7) & (df['video_clicked_mean'] > 0.7)).astype(int)
    df['risky_match'] = ((df['user_clicked_mean'] < 0.3) & (df['video_clicked_mean'] < 0.3)).astype(int)
    
    # ENCODE CATEGORICALS
    df['cat_enc'] = pd.Categorical(df['category']).codes
    df['dev_enc'] = pd.Categorical(df['device']).codes
    df['time_enc'] = pd.Categorical(df['watch_time_of_day']).codes
    
    return df, agg_stats

# Process
train_fe, agg_stats = comprehensive_features(train, is_train=True)
test_fe, _ = comprehensive_features(test, agg_stats, is_train=False)

print(f"✅ COMPREHENSIVE features! Train: {train_fe.shape}, Test: {test_fe.shape}")
```

```python
# CELL 2: Prepare Data with ALL Features
feature_cols = [
    # Basic
    'video_duration', 'watch_time', 'liked', 'commented', 'subscribed_after', 'recommended',
    'watch_percent', 'engagement_score', 'over_watched', 'watch_complete', 'partial_watch', 
    'quick_exit', 'full_watch',
    # Time
    'hour', 'day_of_week', 'day_of_month', 'month', 'is_weekend', 'is_weekday', 
    'is_prime_time', 'is_morning', 'is_late_night',
    # Video categories
    'video_short', 'video_medium', 'video_long', 'video_very_long',
    # Device
    'is_mobile', 'is_desktop', 'is_tv', 'is_tablet',
    # Category
    'is_gaming', 'is_music', 'is_education',
    # User aggregations
    'user_clicked_mean', 'user_clicked_sum', 'user_clicked_count', 'user_clicked_std',
    'user_watch_time_mean', 'user_watch_time_std', 'user_watch_time_min', 
    'user_watch_time_max', 'user_watch_time_median',
    'user_engagement_score_mean', 'user_engagement_score_sum', 'user_engagement_score_max',
    'user_watch_percent_mean', 'user_watch_percent_std', 'user_watch_percent_median',
    'user_liked_sum', 'user_liked_mean', 'user_recommended_mean',
    'user_over_watched_sum', 'user_full_watch_sum',
    # Video aggregations
    'video_clicked_mean', 'video_clicked_sum', 'video_clicked_count', 'video_clicked_std',
    'video_watch_percent_mean', 'video_watch_percent_std', 'video_watch_percent_median',
    'video_engagement_score_mean', 'video_engagement_score_sum',
    'video_liked_sum', 'video_liked_mean', 'video_full_watch_sum',
    # Category/Device/Time aggregations
    'cat_clicked_mean', 'cat_clicked_std', 'cat_watch_time_mean', 'cat_watch_time_median',
    'cat_engagement_score_mean', 'cat_watch_percent_mean',
    'dev_clicked_mean', 'dev_clicked_std', 'dev_watch_time_mean', 'dev_engagement_score_mean',
    'time_click_rate', 'cat_dev_rate', 'cat_time_rate', 'dev_time_rate',
    # Interaction features
    'user_video_affinity', 'user_cat_affinity', 'engagement_vs_user_avg',
    'watch_vs_user_avg', 'watch_vs_video_avg', 'watch_vs_cat_avg',
    'rec_x_user_rate', 'rec_x_video_rate', 'rec_x_cat_rate',
    # Behavior indicators
    'is_active_user', 'is_engaged_user', 'is_binge_watcher', 'is_casual_user',
    'user_consistency', 'is_popular_video', 'is_viral_video', 'is_niche_video',
    'video_quality_score', 'video_engagement_rate', 'perfect_match', 'risky_match',
    # Encoded
    'cat_enc', 'dev_enc', 'time_enc'
]

X = train_fe[feature_cols].fillna(0)
y = train_fe['clicked']
X_test = test_fe[feature_cols].fillna(0)
test_ids = test_fe['id']

print(f"🎯 Training with {len(feature_cols)} FEATURES on {len(X)} samples!")
```

```python
# CELL 3 FIXED: Clean data + Train Models
preds_proba = []
preds_binary = []

# CRITICAL: Clean the data thoroughly
print("🧹 Cleaning data...")
X_clean = X.copy()
X_test_clean = X_test.copy()

# Replace inf with NaN, then fill
X_clean = X_clean.replace([np.inf, -np.inf], np.nan)
X_test_clean = X_test_clean.replace([np.inf, -np.inf], np.nan)

# Fill all NaN with 0
X_clean = X_clean.fillna(0)
X_test_clean = X_test_clean.fillna(0)

# Check for any remaining issues
print(f"X NaN count: {X_clean.isnull().sum().sum()}")
print(f"X_test NaN count: {X_test_clean.isnull().sum().sum()}")
print(f"X shape: {X_clean.shape}, y shape: {y.shape}")

# Model 1: LightGBM #1 (CPU fallback if GPU fails)
print("\n🚀 LightGBM #1...")
try:
    lgb1 = lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
                              min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
                              device='gpu', random_state=42, verbose=-1)
    lgb1.fit(X_clean, y)
except:
    print("   GPU failed, trying CPU...")
    lgb1 = lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
                              min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
                              random_state=42, verbose=-1)
    lgb1.fit(X_clean, y)
p1 = lgb1.predict_proba(X_test_clean)[:, 1]
preds_proba.append(p1)
preds_binary.append((p1 > 0.5).astype(int))
print(f"✅ LGB1: {p1.mean():.3f}")

# Model 2: XGBoost
print("🚀 XGBoost...")
xgb1 = XGBClassifier(n_estimators=600, learning_rate=0.05, max_depth=6, min_child_weight=5,
                     subsample=0.8, colsample_bytree=0.8,
                     tree_method='gpu_hist', predictor='gpu_predictor', random_state=42)
xgb1.fit(X_clean, y)
p2 = xgb1.predict_proba(X_test_clean)[:, 1]
preds_proba.append(p2)
preds_binary.append((p2 > 0.5).astype(int))
print(f"✅ XGB: {p2.mean():.3f}")

# Model 3: CatBoost
print("🚀 CatBoost...")
cat1 = CatBoostClassifier(iterations=500, learning_rate=0.05, depth=6,
                          task_type='GPU', verbose=0, random_state=42)
cat1.fit(X_clean, y)
p3 = cat1.predict_proba(X_test_clean)[:, 1]
preds_proba.append(p3)
preds_binary.append((p3 > 0.5).astype(int))
print(f"✅ CAT: {p3.mean():.3f}")

# Model 4: LightGBM #2
print("🚀 LightGBM #2...")
try:
    lgb2 = lgb.LGBMClassifier(n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
                              min_child_samples=50, subsample=0.85, colsample_bytree=0.85, 
                              device='gpu', random_state=123, verbose=-1)
    lgb2.fit(X_clean, y)
except:
    print("   GPU failed, trying CPU...")
    lgb2 = lgb.LGBMClassifier(n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
                              min_child_samples=50, subsample=0.85, colsample_bytree=0.85, 
                              random_state=123, verbose=-1)
    lgb2.fit(X_clean, y)
p4 = lgb2.predict_proba(X_test_clean)[:, 1]
preds_proba.append(p4)
preds_binary.append((p4 > 0.5).astype(int))
print(f"✅ LGB2: {p4.mean():.3f}")

# Model 5: Random Forest
print("🚀 Random Forest...")
rf1 = RandomForestClassifier(n_estimators=300, max_depth=15, min_samples_split=10, 
                             min_samples_leaf=5, max_features='sqrt', n_jobs=-1, random_state=42)
rf1.fit(X_clean, y)
p5 = rf1.predict_proba(X_test_clean)[:, 1]
preds_proba.append(p5)
preds_binary.append((p5 > 0.5).astype(int))
print(f"✅ RF: {p5.mean():.3f}")

# Model 6: Extra Trees
print("🚀 Extra Trees...")
et1 = ExtraTreesClassifier(n_estimators=300, max_depth=15, min_samples_split=10,
                           max_features='sqrt', n_jobs=-1, random_state=42)
et1.fit(X_clean, y)
p6 = et1.predict_proba(X_test_clean)[:, 1]
preds_proba.append(p6)
preds_binary.append((p6 > 0.5).astype(int))
print(f"✅ ET: {p6.mean():.3f}")

# Model 7: XGBoost #2
print("🚀 XGBoost #2...")
xgb2 = XGBClassifier(n_estimators=800, learning_rate=0.03, max_depth=5,
                     subsample=0.8, colsample_bytree=0.8,
                     tree_method='gpu_hist', predictor='gpu_predictor', random_state=999)
xgb2.fit(X_clean, y)
p7 = xgb2.predict_proba(X_test_clean)[:, 1]
preds_proba.append(p7)
preds_binary.append((p7 > 0.5).astype(int))
print(f"✅ XGB2: {p7.mean():.3f}")

print(f"\n🎉 ALL 7 MODELS TRAINED!")
```

```python
# CELL 4: ADVANCED Ensemble Strategies
pred_stack_proba = np.column_stack(preds_proba)
pred_stack_binary = np.column_stack(preds_binary)

# Strategy 1: Weighted by model type (boosting > forest)
weights = [0.18, 0.18, 0.18, 0.15, 0.10, 0.10, 0.11]  # lgb1, xgb1, cat1, lgb2, rf, et, xgb2
pred_weighted = np.average(pred_stack_proba, weights=weights, axis=1)

# Strategy 2: Simple average
pred_avg = pred_stack_proba.mean(axis=1)

# Strategy 3: Rank averaging (reduces outlier impact)
pred_rank = np.zeros_like(pred_stack_proba)
for i in range(pred_stack_proba.shape[1]):
    pred_rank[:, i] = pd.Series(pred_stack_proba[:, i]).rank(pct=True)
pred_rank_avg = pred_rank.mean(axis=1)

# Strategy 4: Majority voting
pred_vote = (pred_stack_binary.sum(axis=1) >= 4).astype(int)  # 4/7 threshold

# Strategy 5: Super majority
pred_super_vote = (pred_stack_binary.sum(axis=1) >= 5).astype(int)  # 5/7 threshold

# Strategy 6: Best single models
pred_lgb1_single = preds_binary[0]
pred_cat_single = preds_binary[2]

# CREATE ALL SUBMISSIONS
submissions = {
    'sub_weighted.csv': (pred_weighted > 0.5).astype(int),
    'sub_average.csv': (pred_avg > 0.5).astype(int),
    'sub_rank.csv': (pred_rank_avg > 0.5).astype(int),
    'sub_vote_4of7.csv': pred_vote,
    'sub_vote_5of7.csv': pred_super_vote,
    'sub_lgb_single.csv': pred_lgb1_single,
    'sub_cat_single.csv': pred_cat_single,
}

for name, pred in submissions.items():
    pd.DataFrame({'id': test_ids, 'clicked': pred}).to_csv(name, index=False)
    print(f"✅ {name}: Click rate = {pred.mean():.3f}")

print("\n" + "="*60)
print("🏆 SUBMISSION PRIORITY ORDER:")
print("1. sub_weighted.csv        <- START HERE (best ensemble)")
print("2. sub_rank.csv            <- Robust to outliers")
print("3. sub_vote_4of7.csv       <- Majority vote")
print("4. sub_average.csv         <- Simple average")
print("5. sub_cat_single.csv      <- Best single model backup")
print("="*60)
```

```python
"""
Meta-Stacking Full Pipeline
- OOF base model training
- CV-safe target encoding
- Interaction features + frequency encoding
- Meta-stacker: LogisticRegression + MLP (small)
Adjust DATA_PATH, TARGET_COL and ID_COL as needed.
"""

import os
import gc
import time
import warnings
from typing import List, Tuple, Dict

import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, Ridge
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.metrics import roc_auc_score
from sklearn.neural_network import MLPClassifier
import joblib

# Tree libraries
import lightgbm as lgb
import xgboost as xgb
from catboost import CatBoostClassifier

warnings.simplefilter("ignore")
SEED = 2025
np.random.seed(SEED)

# -------- USER CONFIG ----------
DATA_PATH = "train.csv"         # path to train CSV (must contain train + target)
TEST_PATH = "test.csv"          # path to test CSV for final submit (optional)
TARGET_COL = "clicked"           # name of the target column (binary 0/1)
ID_COL = "id"                   # id column name in test (for submission)
N_FOLDS = 5
SHUFFLE = True
STRATIFY = True                 # set False if not classification or no target
# List categorical columns if you know them (for target-encoding and catboost)
CATEGORICALS = ["category", "device", "hour", "is_weekend"]  # example, change accordingly
# --------------------------------

# ------------------ Utilities ------------------
def reduce_mem(df: pd.DataFrame) -> pd.DataFrame:
    # Memory reduction for large CSVs (basic)
    for col in df.select_dtypes(include=["int64"]).columns:
        df[col] = pd.to_numeric(df[col], downcast="integer")
    for col in df.select_dtypes(include=["float64"]).columns:
        df[col] = pd.to_numeric(df[col], downcast="float")
    return df

def safe_check(df: pd.DataFrame) -> pd.DataFrame:
    # replace inf and NaN sensibly
    df = df.replace([np.inf, -np.inf], np.nan)
    # numeric fill: median, categorical: 'missing'
    for c in df.columns:
        if df[c].dtype.kind in "iuf":
            if df[c].isna().any():
                df[c].fillna(df[c].median(), inplace=True)
        else:
            df[c].fillna("__MISSING__", inplace=True)
    return df

# ------------------ Feature Engineering ------------------
def add_basic_features(df: pd.DataFrame) -> pd.DataFrame:
    # Example interactions (tweak to your features)
    # Make sure the columns exist before referencing
    if "user_click_rate" in df.columns and "video_click_rate" in df.columns:
        df["click_rate_prod"] = df["user_click_rate"] * df["video_click_rate"]
    if "hour" in df.columns and "device" in df.columns:
        df["hour_x_device"] = df["hour"].astype(str) + "_" + df["device"].astype(str)
    if "category" in df.columns and "watch_time" in df.columns:
        df["cat_watchtime_ratio"] = df["watch_time"] / (df.groupby("category")["watch_time"].transform("median") + 1e-9)
    # Add simple aggregates
    numeric_cols = df.select_dtypes(include=["int", "float"]).columns.tolist()
    if numeric_cols:
        df["num_feat_mean"] = df[numeric_cols].mean(axis=1)
        df["num_feat_std"] = df[numeric_cols].std(axis=1).fillna(0)
    return df

def frequency_encode(df: pd.DataFrame, cols: List[str]) -> pd.DataFrame:
    for c in cols:
        freq = df[c].value_counts(dropna=False).to_dict()
        df[f"{c}_freq"] = df[c].map(freq).astype(np.int32)
    return df

def target_encode_cv(train: pd.DataFrame, col: str, target: str, n_splits=5, seed=SEED) -> Tuple[np.ndarray, float]:
    """
    K-fold target encoding (out-of-fold mean). Returns OOF array and global mean.
    """
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)
    oof = np.zeros(train.shape[0], dtype=np.float32)
    global_mean = train[target].mean()
    for tr_idx, val_idx in skf.split(train, train[target]):
        tr, val = train.iloc[tr_idx], train.iloc[val_idx]
        means = tr.groupby(col)[target].mean()
        oof[val_idx] = train.loc[val_idx, col].map(means).fillna(global_mean).values
    return oof, global_mean

def apply_target_encoding(train: pd.DataFrame, test: pd.DataFrame, cols: List[str], target: str, n_splits=5):
    # For each categorical column, create TE column by K-fold on train, then map on test using full-train means
    for c in cols:
        print(f"Target encoding {c}")
        oof, global_mean = target_encode_cv(train[[c, target]], col=c, target=target, n_splits=n_splits)
        train[f"{c}_te"] = oof
        # test mapping
        full_means = train.groupby(c)[target].mean()
        test[f"{c}_te"] = test[c].map(full_means).fillna(global_mean).values
    return train, test

# ------------------ OOF training helper ------------------
def get_oof_predictions(clf_name: str,
                        clf,
                        X: pd.DataFrame, y: pd.Series,
                        X_test: pd.DataFrame,
                        folds: StratifiedKFold) -> Tuple[np.ndarray, np.ndarray]:
    n_train = X.shape[0]
    n_test = X_test.shape[0] if X_test is not None else 0
    oof_preds = np.zeros(n_train, dtype=np.float32)
    test_preds = np.zeros(n_test, dtype=np.float32)
    scores = []

    for fold, (tr_idx, val_idx) in enumerate(folds.split(X, y)):
        print(f"[{clf_name}] Fold {fold+1}")
        X_tr, X_val = X.iloc[tr_idx], X.iloc[val_idx]
        y_tr, y_val = y.iloc[tr_idx], y.iloc[val_idx]

        # fit
        if hasattr(clf, "fit"):
            # special handling for xgboost/lightgbm catboost
            if isinstance(clf, lgb.LGBMClassifier):
                clf.fit(X_tr, y_tr,
                        eval_set=[(X_val, y_val)],
                        early_stopping_rounds=50,
                        verbose=False)
                val_pred = clf.predict_proba(X_val)[:, 1]
                test_pred = clf.predict_proba(X_test)[:, 1] if n_test else 0
            elif isinstance(clf, xgb.XGBClassifier):
                clf.fit(X_tr, y_tr,
                        eval_set=[(X_val, y_val)],
                        early_stopping_rounds=50,
                        verbose=False)
                val_pred = clf.predict_proba(X_val)[:, 1]
                test_pred = clf.predict_proba(X_test)[:, 1] if n_test else 0
            elif isinstance(clf, CatBoostClassifier):
                clf.fit(X_tr, y_tr,
                        eval_set=(X_val, y_val),
                        verbose=False,
                        early_stopping_rounds=50)
                val_pred = clf.predict_proba(X_val)[:, 1]
                test_pred = clf.predict_proba(X_test)[:, 1] if n_test else 0
            else:
                clf.fit(X_tr, y_tr)
                if hasattr(clf, "predict_proba"):
                    val_pred = clf.predict_proba(X_val)[:, 1]
                    test_pred = clf.predict_proba(X_test)[:, 1] if n_test else 0
                else:
                    val_pred = clf.predict(X_val)
                    test_pred = clf.predict(X_test) if n_test else 0
        else:
            raise ValueError("Classifier has no fit method")

        oof_preds[val_idx] = val_pred
        if n_test:
            test_preds += test_pred / folds.n_splits

        score = roc_auc_score(y_val, val_pred)
        scores.append(score)
        print(f"  Fold {fold+1} AUC: {score:.5f}")

    print(f"[{clf_name}] Mean AUC: {np.mean(scores):.5f}")
    return oof_preds, test_preds

# ------------------ Model definitions ------------------
def make_base_models(seed=SEED):
    # Provide reasonable GPU-friendly parameters for LightGBM/XGBoost/CatBoost
    lgb_params = {
        "n_estimators": 1000, "learning_rate": 0.05, "num_leaves": 64,
        "colsample_bytree": 0.6, "subsample": 0.8, "random_state": seed
    }
    xgb_params = {"n_estimators": 1000, "learning_rate": 0.05, "max_depth": 6, "verbosity": 0, "random_state": seed}
    cat_params = {"iterations": 1000, "learning_rate": 0.05, "verbose": False, "random_seed": seed}

    models = {
        "lgb": lgb.LGBMClassifier(**lgb_params),
        "xgb": xgb.XGBClassifier(**xgb_params, use_label_encoder=False),
        "cat": CatBoostClassifier(**cat_params),
        "rf": RandomForestClassifier(n_estimators=400, n_jobs=-1, random_state=seed),
        "et": ExtraTreesClassifier(n_estimators=300, n_jobs=-1, random_state=seed),
        "ridge": Ridge(alpha=1.0),  # regression used as base (prob outputs via decision function later)
        "logreg": LogisticRegression(max_iter=1000, solver="lbfgs")
    }
    return models

# ------------------ Main pipeline ------------------
def main():
    # Load data
    print("Loading data...")
    train = pd.read_csv(DATA_PATH)
    test = pd.read_csv(TEST_PATH) if os.path.exists(TEST_PATH) else None

    # Basic cleaning
    train = reduce_mem(train)
    train = safe_check(train)
    if test is not None:
        test = reduce_mem(test)
        test = safe_check(test)

    features = [c for c in train.columns if c not in [TARGET_COL]]
    if ID_COL in features:
        features.remove(ID_COL)

    # Add derived features (interactions)
    print("Adding derived features...")
    train = add_basic_features(train)
    if test is not None:
        test = add_basic_features(test)

    # Frequency encoding for categorical columns (fast)
    print("Frequency encoding...")
    freq_cols = [c for c in CATEGORICALS if c in train.columns]
    train = frequency_encode(train, freq_cols)
    if test is not None:
        test = frequency_encode(test, freq_cols)

    # Target encoding with K-fold safety
    te_cols = [c for c in CATEGORICALS if c in train.columns]
    if te_cols:
        print("Applying K-fold target encoding...")
        train, test = apply_target_encoding(train, test if test is not None else pd.DataFrame(), te_cols, TARGET_COL, n_splits=N_FOLDS)

    # Rebuild feature list after feature engineering
    exclude_cols = [TARGET_COL]
    if ID_COL in train.columns:
        exclude_cols.append(ID_COL)
    X_cols = [c for c in train.columns if c not in exclude_cols]

    # If test was provided, ensure same features
    if test is not None:
        for c in X_cols:
            if c not in test.columns:
                test[c] = 0

    # Prepare arrays
    X = train[X_cols].copy()
    y = train[TARGET_COL].copy()
    X_test = test[X_cols].copy() if test is not None else None

    # Standard scaler for linear models / MLP
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
    X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=X.columns) if X_test is not None else None

    # Create folds
    folds = StratifiedKFold(n_splits=N_FOLDS, shuffle=SHUFFLE, random_state=SEED) if STRATIFY else StratifiedKFold(n_splits=N_FOLDS, shuffle=SHUFFLE, random_state=SEED)

    # Build base models
    models = make_base_models(seed=SEED)

    # Storage for OOF and test preds for stacking
    oof_df = pd.DataFrame(index=train.index)
    test_df = pd.DataFrame(index=test.index) if test is not None else None

    # Train each base model and collect OOF
    for name, model in models.items():
        print(f"\nTraining base model: {name}")
        # Choose feature set depending on model type (tree vs linear)
        if name in ["ridge", "logreg"]:
            _X = X_scaled
            _X_test = X_test_scaled if X_test is not None else None
        else:
            _X = X
            _X_test = X_test if X_test is not None else None

        oof_preds, test_preds = get_oof_predictions(name, model, _X, y, _X_test, folds)
        oof_df[f"{name}_oof"] = oof_preds
        if test_df is not None:
            test_df[f"{name}_pred"] = test_preds

        # cleanup
        gc.collect()

    # Evaluate base ensemble (simple average) (optional)
    print("\nBase models AUCs on OOF (quick check):")
    for c in oof_df.columns:
        try:
            print(c, roc_auc_score(y, oof_df[c]))
        except Exception:
            pass

    # Save OOF matrix for debug
    oof_df.to_csv("oof_base_preds.csv", index=False)

    # --------------- Meta-stacker training ----------------
    print("\nTraining meta-stacker...")
    # Prepare meta features: base OOF columns + maybe some original strong features
    meta_X = oof_df.copy()
    # Optionally add a couple of strong original features (example)
    extra_for_meta = []
    for c in ["num_feat_mean", "user_click_rate"]:
        if c in X.columns:
            meta_X[c] = X[c]
            extra_for_meta.append(c)

    # Standardize meta features for MLP
    meta_scaler = StandardScaler()
    meta_X_scaled = meta_scaler.fit_transform(meta_X)
    meta_test = test_df.copy() if test_df is not None else None
    if meta_test is not None:
        for c in extra_for_meta:
            if c in X_test.columns:
                meta_test[c] = X_test[c]
        meta_test_scaled = meta_scaler.transform(meta_test)
    else:
        meta_test_scaled = None

    # Meta model 1: Logistic Regression
    lr = LogisticRegression(max_iter=2000)
    # Meta model 2: simple MLP
    mlp = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=1000, random_state=SEED)

    # Cross-validated meta OOF
    meta_folds = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)
    meta_oof_lr = np.zeros(meta_X.shape[0])
    meta_oof_mlp = np.zeros(meta_X.shape[0])
    meta_test_preds_lr = np.zeros(meta_test.shape[0]) if meta_test is not None else None
    meta_test_preds_mlp = np.zeros(meta_test.shape[0]) if meta_test is not None else None

    for fold, (tr_idx, val_idx) in enumerate(meta_folds.split(meta_X_scaled, y)):
        print(f"Meta fold {fold+1}")
        Xtr_meta, Xval_meta = meta_X_scaled[tr_idx], meta_X_scaled[val_idx]
        ytr_meta, yval_meta = y.iloc[tr_idx], y.iloc[val_idx]

        lr.fit(Xtr_meta, ytr_meta)
        meta_oof_lr[val_idx] = lr.predict_proba(Xval_meta)[:, 1]
        if meta_test is not None:
            meta_test_preds_lr += lr.predict_proba(meta_test_scaled)[:, 1] / meta_folds.n_splits

        mlp.fit(Xtr_meta, ytr_meta)
        meta_oof_mlp[val_idx] = mlp.predict_proba(Xval_meta)[:, 1]
        if meta_test is not None:
            meta_test_preds_mlp += mlp.predict_proba(meta_test_scaled)[:, 1] / meta_folds.n_splits

    print("Meta LR OOF AUC:", roc_auc_score(y, meta_oof_lr))
    print("Meta MLP OOF AUC:", roc_auc_score(y, meta_oof_mlp))

    # Combine meta predictions (simple average or weighted)
    meta_oof_stack = (meta_oof_lr + meta_oof_mlp) / 2.0
    if meta_test is not None:
        meta_test_stack = (meta_test_preds_lr + meta_test_preds_mlp) / 2.0
    else:
        meta_test_stack = None

    print("Stacked Meta OOF AUC:", roc_auc_score(y, meta_oof_stack))

    # Final step: Save meta models and produce final submission
    joblib.dump(lr, "meta_lr.joblib")
    joblib.dump(mlp, "meta_mlp.joblib")
    oof_df["meta_oof"] = meta_oof_stack
    oof_df.to_csv("final_oof_stack.csv", index=False)

    if meta_test_stack is not None:
        submission = pd.DataFrame({ID_COL: test[ID_COL], TARGET_COL: meta_test_stack})
        submission.to_csv("submission.csv", index=False)
        print("Submission saved: submission.csv")

    print("Done.")

if __name__ == "__main__":
    start = time.time()
    main()
    print("Total elapsed:", time.time() - start)
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')

# Load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(f"Train: {train.shape}, Test: {test.shape}")
print(f"Click rate: {train['clicked'].mean():.3f}")


def create_basic_features(df):
    """Create features that don't require aggregations (no leakage)."""
    df = df.copy()
    
    # Fix data types
    df['liked'] = df['liked'].replace({'yes': 1, 'no': 0, 'Yes': 1, 'No': 0})
    df['liked'] = pd.to_numeric(df['liked'], errors='coerce').fillna(0)
    df['commented'] = df['commented'].fillna(0).astype(int)
    df['subscribed_after'] = df['subscribed_after'].fillna(0).astype(int)
    
    # Basic engagement
    df['watch_percent'] = (df['watch_time'] / df['video_duration']).fillna(0).clip(0, 5)
    df['engagement_score'] = df['liked'] + df['commented']*2 + df['subscribed_after']*3
    df['over_watched'] = (df['watch_time'] > df['video_duration']).astype(int)
    df['watch_complete'] = np.minimum(df['watch_percent'], 1.0)
    df['partial_watch'] = ((df['watch_percent'] > 0.3) & (df['watch_percent'] < 0.9)).astype(int)
    df['quick_exit'] = (df['watch_percent'] < 0.1).astype(int)
    df['full_watch'] = (df['watch_percent'] >= 0.95).astype(int)
    
    # Time features
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    df['hour'] = df['timestamp'].dt.hour
    df['day_of_week'] = df['timestamp'].dt.dayofweek
    df['day_of_month'] = df['timestamp'].dt.day
    df['month'] = df['timestamp'].dt.month
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_weekday'] = (df['day_of_week'] < 5).astype(int)
    df['is_prime_time'] = df['hour'].isin([18,19,20,21,22]).astype(int)
    df['is_morning'] = df['hour'].isin([6,7,8,9,10]).astype(int)
    df['is_late_night'] = df['hour'].isin([0,1,2,3,4,5]).astype(int)
    
    # Create time_of_day feature
    df['time_of_day'] = pd.cut(df['hour'], bins=[0,6,12,18,24], labels=['night','morning','afternoon','evening'], include_lowest=True)
    
    # Video length
    df['video_short'] = (df['video_duration'] < 300).astype(int)
    df['video_medium'] = ((df['video_duration'] >= 300) & (df['video_duration'] < 1200)).astype(int)
    df['video_long'] = ((df['video_duration'] >= 1200) & (df['video_duration'] < 3000)).astype(int)
    df['video_very_long'] = (df['video_duration'] >= 3000).astype(int)
    
    # Device features
    df['is_mobile'] = (df['device'] == 'Mobile').astype(int)
    df['is_desktop'] = (df['device'] == 'Desktop').astype(int)
    df['is_tv'] = (df['device'] == 'TV').astype(int)
    df['is_tablet'] = (df['device'] == 'Tablet').astype(int)
    
    # Category features
    df['is_gaming'] = (df['category'] == 'Gaming').astype(int)
    df['is_music'] = (df['category'] == 'Music').astype(int)
    df['is_education'] = (df['category'] == 'Education').astype(int)
    
    # Encode categoricals
    df['cat_enc'] = pd.Categorical(df['category']).codes
    df['dev_enc'] = pd.Categorical(df['device']).codes
    df['time_enc'] = pd.Categorical(df['time_of_day']).codes
    
    return df


def add_aggregation_features(train_fold, target_df, global_mean):
    """
    Compute aggregations on train_fold ONLY, apply to target_df.
    This prevents leakage.
    """
    # USER AGGREGATIONS
    user_stats = train_fold.groupby('user_id').agg({
        'clicked': ['mean', 'sum', 'count', 'std'],
        'watch_time': ['mean', 'std', 'median'],
        'engagement_score': ['mean', 'sum'],
        'watch_percent': ['mean', 'std'],
        'liked': ['sum', 'mean'],
        'full_watch': 'sum'
    })
    user_stats.columns = ['_'.join(col).strip() for col in user_stats.columns]
    user_stats = user_stats.add_prefix('user_')
    
    # VIDEO AGGREGATIONS
    video_stats = train_fold.groupby('video_id').agg({
        'clicked': ['mean', 'sum', 'count', 'std'],
        'watch_percent': ['mean', 'median'],
        'engagement_score': 'mean',
        'liked': ['sum', 'mean']
    })
    video_stats.columns = ['_'.join(col).strip() for col in video_stats.columns]
    video_stats = video_stats.add_prefix('video_')
    
    # CATEGORY AGGREGATIONS
    cat_stats = train_fold.groupby('category').agg({
        'clicked': ['mean', 'std'],
        'watch_time': 'mean',
        'engagement_score': 'mean'
    })
    cat_stats.columns = ['_'.join(col).strip() for col in cat_stats.columns]
    cat_stats = cat_stats.add_prefix('cat_')
    
    # DEVICE AGGREGATIONS
    dev_stats = train_fold.groupby('device').agg({
        'clicked': ['mean', 'std'],
        'watch_time': 'mean'
    })
    dev_stats.columns = ['_'.join(col).strip() for col in dev_stats.columns]
    dev_stats = dev_stats.add_prefix('dev_')
    
    # HOUR AGGREGATIONS (time patterns)
    hour_stats = train_fold.groupby('hour')['clicked'].mean().to_frame('hour_click_rate')
    
    # Merge to target dataframe
    target_df = target_df.merge(user_stats, left_on='user_id', right_index=True, how='left')
    target_df = target_df.merge(video_stats, left_on='video_id', right_index=True, how='left')
    target_df = target_df.merge(cat_stats, left_on='category', right_index=True, how='left')
    target_df = target_df.merge(dev_stats, left_on='device', right_index=True, how='left')
    target_df = target_df.merge(hour_stats, left_on='hour', right_index=True, how='left')
    
    # Fill missing with global mean/0
    for col in target_df.columns:
        if target_df[col].dtype in ['float64', 'int64'] and target_df[col].isnull().any():
            if 'mean' in col or 'rate' in col:
                target_df[col] = target_df[col].fillna(global_mean)
            else:
                target_df[col] = target_df[col].fillna(0)
    
    return target_df


# Create basic features (no leakage)
print("Creating basic features...")
train_fe = create_basic_features(train)
test_fe = create_basic_features(test)

# Stratified K-Fold Cross-Validation
n_folds = 5
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

# Initialize prediction arrays for each model
model_names = ['lgb1', 'xgb1', 'cat1', 'lgb2', 'rf', 'et', 'xgb2']
oof_preds = {name: np.zeros(len(train_fe)) for name in model_names}
test_preds = {name: np.zeros(len(test_fe)) for name in model_names}
cv_scores = {name: [] for name in model_names}

global_mean = train_fe['clicked'].mean()

print(f"\n{'='*60}")
print(f"Starting {n_folds}-Fold Cross-Validation with 7 Models")
print(f"{'='*60}\n")

for fold, (train_idx, val_idx) in enumerate(skf.split(train_fe, train_fe['clicked']), 1):
    print(f"\n{'='*50}")
    print(f"FOLD {fold}/{n_folds}")
    print(f"{'='*50}")
    
    # Split data
    train_fold = train_fe.iloc[train_idx].copy()
    val_fold = train_fe.iloc[val_idx].copy()
    test_fold = test_fe.copy()
    
    # Add aggregation features (computed ONLY on train_fold)
    train_fold_agg = add_aggregation_features(train_fold, train_fold.copy(), global_mean)
    val_fold_agg = add_aggregation_features(train_fold, val_fold, global_mean)
    test_fold_agg = add_aggregation_features(train_fold, test_fold, global_mean)
    
    # Get feature columns - exclude all non-numeric columns
    exclude_cols = ['clicked', 'id', 'user_id', 'video_id', 'category', 'device', 'timestamp', 
                    'time_of_day', 'watch_time_of_day']
    
    # Only select numeric columns
    feature_cols = []
    for c in train_fold_agg.columns:
        if c not in exclude_cols and train_fold_agg[c].dtype in ['int64', 'int32', 'float64', 'float32', 'int8', 'int16', 'uint8', 'uint16', 'bool']:
            feature_cols.append(c)
    
    # Prepare data
    X_train = train_fold_agg[feature_cols].fillna(0)
    y_train = train_fold_agg['clicked']
    X_val = val_fold_agg[feature_cols].fillna(0)
    y_val = val_fold_agg['clicked']
    X_test = test_fold_agg[feature_cols].fillna(0)
    
    # Clean data
    X_train = X_train.replace([np.inf, -np.inf], 0)
    X_val = X_val.replace([np.inf, -np.inf], 0)
    X_test = X_test.replace([np.inf, -np.inf], 0)
    
    print(f"Train: {X_train.shape}, Val: {X_val.shape}, Features: {len(feature_cols)}")
    
    # Model 1: LightGBM #1
    print("\n🚀 Training LightGBM #1...")
    try:
        lgb1 = lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
                                  min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
                                  device='gpu', random_state=42, verbose=-1)
        lgb1.fit(X_train, y_train)
    except:
        lgb1 = lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
                                  min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
                                  random_state=42, verbose=-1)
        lgb1.fit(X_train, y_train)
    
    val_pred = lgb1.predict_proba(X_val)[:, 1]
    test_pred = lgb1.predict_proba(X_test)[:, 1]
    oof_preds['lgb1'][val_idx] = val_pred
    test_preds['lgb1'] += test_pred / n_folds
    score = roc_auc_score(y_val, val_pred)
    cv_scores['lgb1'].append(score)
    print(f"   LGB1 AUC: {score:.5f}")
    
    # Model 2: XGBoost #1
    print("🚀 Training XGBoost #1...")
    xgb1 = XGBClassifier(n_estimators=600, learning_rate=0.05, max_depth=6, min_child_weight=5,
                         subsample=0.8, colsample_bytree=0.8,
                         tree_method='gpu_hist', predictor='gpu_predictor', random_state=42, eval_metric='logloss')
    xgb1.fit(X_train, y_train)
    val_pred = xgb1.predict_proba(X_val)[:, 1]
    test_pred = xgb1.predict_proba(X_test)[:, 1]
    oof_preds['xgb1'][val_idx] = val_pred
    test_preds['xgb1'] += test_pred / n_folds
    score = roc_auc_score(y_val, val_pred)
    cv_scores['xgb1'].append(score)
    print(f"   XGB1 AUC: {score:.5f}")
    
    # Model 3: CatBoost
    print("🚀 Training CatBoost...")
    cat1 = CatBoostClassifier(iterations=500, learning_rate=0.05, depth=6,
                              task_type='GPU', verbose=0, random_state=42)
    cat1.fit(X_train, y_train)
    val_pred = cat1.predict_proba(X_val)[:, 1]
    test_pred = cat1.predict_proba(X_test)[:, 1]
    oof_preds['cat1'][val_idx] = val_pred
    test_preds['cat1'] += test_pred / n_folds
    score = roc_auc_score(y_val, val_pred)
    cv_scores['cat1'].append(score)
    print(f"   CAT AUC: {score:.5f}")
    
    # Model 4: LightGBM #2
    print("🚀 Training LightGBM #2...")
    try:
        lgb2 = lgb.LGBMClassifier(n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
                                  min_child_samples=50, subsample=0.85, colsample_bytree=0.85,
                                  device='gpu', random_state=123, verbose=-1)
        lgb2.fit(X_train, y_train)
    except:
        lgb2 = lgb.LGBMClassifier(n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
                                  min_child_samples=50, subsample=0.85, colsample_bytree=0.85,
                                  random_state=123, verbose=-1)
        lgb2.fit(X_train, y_train)
    val_pred = lgb2.predict_proba(X_val)[:, 1]
    test_pred = lgb2.predict_proba(X_test)[:, 1]
    oof_preds['lgb2'][val_idx] = val_pred
    test_preds['lgb2'] += test_pred / n_folds
    score = roc_auc_score(y_val, val_pred)
    cv_scores['lgb2'].append(score)
    print(f"   LGB2 AUC: {score:.5f}")
    
    # Model 5: Random Forest
    print("🚀 Training Random Forest...")
    rf1 = RandomForestClassifier(n_estimators=300, max_depth=15, min_samples_split=10,
                                 min_samples_leaf=5, max_features='sqrt', n_jobs=-1, random_state=42)
    rf1.fit(X_train, y_train)
    val_pred = rf1.predict_proba(X_val)[:, 1]
    test_pred = rf1.predict_proba(X_test)[:, 1]
    oof_preds['rf'][val_idx] = val_pred
    test_preds['rf'] += test_pred / n_folds
    score = roc_auc_score(y_val, val_pred)
    cv_scores['rf'].append(score)
    print(f"   RF AUC: {score:.5f}")
    
    # Model 6: Extra Trees
    print("🚀 Training Extra Trees...")
    et1 = ExtraTreesClassifier(n_estimators=300, max_depth=15, min_samples_split=10,
                               max_features='sqrt', n_jobs=-1, random_state=42)
    et1.fit(X_train, y_train)
    val_pred = et1.predict_proba(X_val)[:, 1]
    test_pred = et1.predict_proba(X_test)[:, 1]
    oof_preds['et'][val_idx] = val_pred
    test_preds['et'] += test_pred / n_folds
    score = roc_auc_score(y_val, val_pred)
    cv_scores['et'].append(score)
    print(f"   ET AUC: {score:.5f}")
    
    # Model 7: XGBoost #2
    print("🚀 Training XGBoost #2...")
    xgb2 = XGBClassifier(n_estimators=800, learning_rate=0.03, max_depth=5,
                         subsample=0.8, colsample_bytree=0.8,
                         tree_method='gpu_hist', predictor='gpu_predictor', random_state=999, eval_metric='logloss')
    xgb2.fit(X_train, y_train)
    val_pred = xgb2.predict_proba(X_val)[:, 1]
    test_pred = xgb2.predict_proba(X_test)[:, 1]
    oof_preds['xgb2'][val_idx] = val_pred
    test_preds['xgb2'] += test_pred / n_folds
    score = roc_auc_score(y_val, val_pred)
    cv_scores['xgb2'].append(score)
    print(f"   XGB2 AUC: {score:.5f}")

# Print CV Results
print(f"\n{'='*60}")
print("CROSS-VALIDATION RESULTS:")
print(f"{'='*60}")
for name in model_names:
    oof_score = roc_auc_score(train_fe['clicked'], oof_preds[name])
    mean_cv = np.mean(cv_scores[name])
    std_cv = np.std(cv_scores[name])
    print(f"{name.upper():10s} | OOF AUC: {oof_score:.5f} | CV: {mean_cv:.5f} (+/- {std_cv:.5f})")

# Ensemble Strategies
print(f"\n{'='*60}")
print("CREATING ENSEMBLE SUBMISSIONS:")
print(f"{'='*60}\n")

# Stack predictions
test_stack = np.column_stack([test_preds[name] for name in model_names])

# Strategy 1: Weighted average (boosting > forests)
weights = [0.18, 0.18, 0.18, 0.15, 0.10, 0.10, 0.11]
pred_weighted = np.average(test_stack, weights=weights, axis=1)

# Strategy 2: Simple average
pred_avg = test_stack.mean(axis=1)

# Strategy 3: Rank averaging
pred_rank = np.zeros_like(test_stack)
for i in range(test_stack.shape[1]):
    pred_rank[:, i] = pd.Series(test_stack[:, i]).rank(pct=True)
pred_rank_avg = pred_rank.mean(axis=1)

# Strategy 4: Best single models
pred_lgb1 = test_preds['lgb1']
pred_cat1 = test_preds['cat1']
pred_xgb1 = test_preds['xgb1']

# Create submissions
submissions = {
    'sub_weighted_cv.csv': (pred_weighted > 0.5).astype(int),
    'sub_average_cv.csv': (pred_avg > 0.5).astype(int),
    'sub_rank_cv.csv': (pred_rank_avg > 0.5).astype(int),
    'sub_lgb1_cv.csv': (pred_lgb1 > 0.5).astype(int),
    'sub_cat1_cv.csv': (pred_cat1 > 0.5).astype(int),
    'sub_xgb1_cv.csv': (pred_xgb1 > 0.5).astype(int),
}

for name, pred in submissions.items():
    pd.DataFrame({'id': test_fe['id'], 'clicked': pred}).to_csv(name, index=False)
    print(f"✅ {name}: Click rate = {pred.mean():.3f}")

print("\n" + "="*60)
print("🏆 SUBMISSION PRIORITY (try in order):")
print("1. sub_weighted_cv.csv    <- Weighted ensemble")
print("2. sub_rank_cv.csv        <- Rank averaging") 
print("3. sub_average_cv.csv     <- Simple average")
print("4. Best single model based on CV scores above")
print("="*60)
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from scipy.stats import spearmanr
import warnings
warnings.filterwarnings('ignore')

# ===== CUDA AVAILABILITY CHECK =====
import torch
print("="*70)
print("🔍 CUDA AVAILABILITY CHECK")
print("="*70)
print(f"PyTorch CUDA: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Device: {torch.cuda.get_device_name(0)}")
    print(f"CUDA version: {torch.version.cuda}")

# Check XGBoost GPU
xgb_gpu = False
try:
    import xgboost as xgb_test
    dtest = xgb_test.DMatrix([[1,2]], label=[0])
    xgb_test.train({'tree_method': 'gpu_hist'}, dtest, num_boost_round=1)
    xgb_gpu = True
    print(f"XGBoost GPU: Available")
except:
    print(f"XGBoost GPU: Not available (will use CPU hist)")

use_gpu = torch.cuda.is_available()
print(f"LightGBM GPU: Will attempt")
print(f"CatBoost GPU: Will attempt")
print("="*70 + "\n")

print("="*70)
print("🚀 ULTIMATE ML PIPELINE - CUDA ENABLED + Proper CV + Neural Nets")
print("="*70)

# Load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(f"\n📊 Data Loaded:")
print(f"   Train: {train.shape}, Test: {test.shape}")
print(f"   Click rate: {train['clicked'].mean():.3f}")
print(f"   Columns: {list(train.columns)}")

# CRITICAL CHECK: Remove watch_percent if it exists (it's leakage!)
if 'watch_percent' in train.columns:
    print(f"   ⚠️  WARNING: Removing 'watch_percent' from original data (potential leakage)")
    train = train.drop('watch_percent', axis=1)
    test = test.drop('watch_percent', axis=1) if 'watch_percent' in test.columns else test


def create_comprehensive_features(df):
    """Create ALL features - basic + advanced."""
    df = df.copy()
    
    # ===== FIX DATA TYPES =====
    if df['liked'].dtype == 'object':
        df['liked'] = df['liked'].str.lower().replace({'yes': 1, 'no': 0})
    df['liked'] = pd.to_numeric(df['liked'], errors='coerce').fillna(0).astype(int)
    df['commented'] = pd.to_numeric(df['commented'], errors='coerce').fillna(0).astype(int)
    df['subscribed_after'] = pd.to_numeric(df['subscribed_after'], errors='coerce').fillna(0).astype(int)
    df['recommended'] = pd.to_numeric(df['recommended'], errors='coerce').fillna(0).astype(int)
    
    # CRITICAL FIX: Ensure clicked is binary (0 or 1 only)
    if 'clicked' in df.columns:
        df['clicked'] = pd.to_numeric(df['clicked'], errors='coerce').fillna(0).astype(int)
        df['clicked'] = df['clicked'].clip(0, 1)  # Force to be 0 or 1
        print(f"   ✓ Clicked column: {df['clicked'].unique()}, rate: {df['clicked'].mean():.3f}")
    
    # ===== BASIC ENGAGEMENT FEATURES =====
    # NOTE: liked, commented, subscribed_after happen AFTER the click
    # So we can't use them directly - but we CAN use user/video aggregations of them
    df['watch_percent'] = (df['watch_time'] / (df['video_duration'] + 1)).fillna(0).clip(0, 5)
    
    # REMOVED: Don't use post-click features directly
    # df['engagement_score'] = df['liked'] + df['commented']*2 + df['subscribed_after']*3
    
    df['over_watched'] = (df['watch_time'] > df['video_duration']).astype(int)
    df['watch_complete'] = np.minimum(df['watch_percent'], 1.0)
    df['partial_watch'] = ((df['watch_percent'] > 0.3) & (df['watch_percent'] < 0.9)).astype(int)
    df['quick_exit'] = (df['watch_percent'] < 0.1).astype(int)
    df['full_watch'] = (df['watch_percent'] >= 0.95).astype(int)
    df['half_watch'] = ((df['watch_percent'] >= 0.45) & (df['watch_percent'] <= 0.55)).astype(int)
    
    # Watch time features
    df['watch_time_log'] = np.log1p(df['watch_time'])
    df['video_duration_log'] = np.log1p(df['video_duration'])
    df['watch_speed'] = df['watch_time'] / (df['video_duration'] + 1)
    
    # ===== TIME FEATURES =====
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    df['hour'] = df['timestamp'].dt.hour
    df['day_of_week'] = df['timestamp'].dt.dayofweek
    df['day_of_month'] = df['timestamp'].dt.day
    df['month'] = df['timestamp'].dt.month
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_weekday'] = (df['day_of_week'] < 5).astype(int)
    df['is_prime_time'] = df['hour'].isin([18,19,20,21,22]).astype(int)
    df['is_morning'] = df['hour'].isin([6,7,8,9,10]).astype(int)
    df['is_afternoon'] = df['hour'].isin([12,13,14,15,16,17]).astype(int)
    df['is_late_night'] = df['hour'].isin([0,1,2,3,4,5]).astype(int)
    
    # ===== VIDEO LENGTH CATEGORIES =====
    df['video_short'] = (df['video_duration'] < 300).astype(int)
    df['video_medium'] = ((df['video_duration'] >= 300) & (df['video_duration'] < 1200)).astype(int)
    df['video_long'] = ((df['video_duration'] >= 1200) & (df['video_duration'] < 3000)).astype(int)
    df['video_very_long'] = (df['video_duration'] >= 3000).astype(int)
    
    # ===== DEVICE FEATURES =====
    df['is_mobile'] = (df['device'] == 'Mobile').astype(int)
    df['is_desktop'] = (df['device'] == 'Desktop').astype(int)
    df['is_tv'] = (df['device'] == 'TV').astype(int)
    df['is_tablet'] = (df['device'] == 'Tablet').astype(int)
    
    # ===== CATEGORY FEATURES =====
    for cat in ['Gaming', 'Music', 'Education', 'Sports', 'Comedy', 'News', 'Tech', 'Lifestyle']:
        df[f'is_{cat.lower()}'] = (df['category'] == cat).astype(int)
    
    # ===== WATCH_TIME_OF_DAY FEATURES =====
    if 'watch_time_of_day' in df.columns:
        for tod in ['Morning', 'Afternoon', 'Evening', 'Night']:
            df[f'is_{tod.lower()}_watch'] = (df['watch_time_of_day'] == tod).astype(int)
    
    # ===== INTERACTION FEATURES =====
    df['liked_and_watched'] = df['liked'] * df['watch_complete']
    df['rec_and_watched'] = df['recommended'] * df['watch_complete']
    df['engagement_per_second'] = df['engagement_score'] / (df['watch_time'] + 1)
    
    # ===== ENCODE CATEGORICALS =====
    df['cat_enc'] = pd.Categorical(df['category']).codes
    df['dev_enc'] = pd.Categorical(df['device']).codes
    if 'watch_time_of_day' in df.columns:
        df['time_of_day_enc'] = pd.Categorical(df['watch_time_of_day']).codes
    
    return df


def add_aggregation_features(train_fold, target_df, global_mean):
    """Compute aggregations on train_fold ONLY, apply to target_df."""
    
    # USER AGGREGATIONS
    user_stats = train_fold.groupby('user_id').agg({
        'clicked': ['mean', 'sum', 'count', 'std'],
        'watch_time': ['mean', 'std', 'median', 'max'],
        'engagement_score': ['mean', 'sum', 'max'],
        'watch_percent': ['mean', 'std', 'median'],
        'liked': ['sum', 'mean'],
        'full_watch': ['sum', 'mean'],
        'recommended': 'mean',
        'over_watched': 'sum'
    })
    user_stats.columns = ['_'.join(col).strip() for col in user_stats.columns]
    user_stats = user_stats.add_prefix('user_')
    
    # VIDEO AGGREGATIONS
    video_stats = train_fold.groupby('video_id').agg({
        'clicked': ['mean', 'sum', 'count', 'std'],
        'watch_percent': ['mean', 'std', 'median'],
        'engagement_score': ['mean', 'sum'],
        'liked': ['sum', 'mean'],
        'full_watch': ['sum', 'mean']
    })
    video_stats.columns = ['_'.join(col).strip() for col in video_stats.columns]
    video_stats = video_stats.add_prefix('video_')
    
    # CATEGORY AGGREGATIONS
    cat_stats = train_fold.groupby('category').agg({
        'clicked': ['mean', 'std'],
        'watch_time': ['mean', 'median'],
        'engagement_score': 'mean',
        'watch_percent': 'mean'
    })
    cat_stats.columns = ['_'.join(col).strip() for col in cat_stats.columns]
    cat_stats = cat_stats.add_prefix('cat_')
    
    # DEVICE AGGREGATIONS
    dev_stats = train_fold.groupby('device').agg({
        'clicked': ['mean', 'std'],
        'watch_time': 'mean',
        'engagement_score': 'mean'
    })
    dev_stats.columns = ['_'.join(col).strip() for col in dev_stats.columns]
    dev_stats = dev_stats.add_prefix('dev_')
    
    # TIME OF DAY AGGREGATIONS
    if 'watch_time_of_day' in train_fold.columns:
        time_stats = train_fold.groupby('watch_time_of_day')['clicked'].mean().to_frame('time_click_rate')
    
    # HOUR AGGREGATIONS
    hour_stats = train_fold.groupby('hour')['clicked'].mean().to_frame('hour_click_rate')
    
    # COMBO AGGREGATIONS
    cat_dev_stats = train_fold.groupby(['category', 'device'])['clicked'].mean().reset_index()
    cat_dev_stats.columns = ['category', 'device', 'cat_dev_rate']
    
    # Merge to target
    target_df = target_df.merge(user_stats, left_on='user_id', right_index=True, how='left')
    target_df = target_df.merge(video_stats, left_on='video_id', right_index=True, how='left')
    target_df = target_df.merge(cat_stats, left_on='category', right_index=True, how='left')
    target_df = target_df.merge(dev_stats, left_on='device', right_index=True, how='left')
    if 'watch_time_of_day' in train_fold.columns:
        target_df = target_df.merge(time_stats, left_on='watch_time_of_day', right_index=True, how='left')
    target_df = target_df.merge(hour_stats, left_on='hour', right_index=True, how='left')
    target_df = target_df.merge(cat_dev_stats, on=['category', 'device'], how='left')
    
    # ADVANCED INTERACTION FEATURES
    target_df['user_video_affinity'] = target_df['user_clicked_mean'] * target_df['video_clicked_mean']
    target_df['user_cat_affinity'] = target_df['user_clicked_mean'] * target_df['cat_clicked_mean']
    target_df['watch_vs_user_avg'] = target_df['watch_percent'] / (target_df['user_watch_percent_mean'] + 0.01)
    target_df['watch_vs_video_avg'] = target_df['watch_percent'] / (target_df['video_watch_percent_mean'] + 0.01)
    target_df['engagement_vs_user_avg'] = target_df['engagement_score'] / (target_df['user_engagement_score_mean'] + 0.01)
    target_df['is_active_user'] = (target_df['user_clicked_count'] > target_df['user_clicked_count'].median()).astype(int)
    target_df['is_popular_video'] = (target_df['video_clicked_count'] > target_df['video_clicked_count'].median()).astype(int)
    
    # Fill missing
    for col in target_df.columns:
        if target_df[col].dtype in ['float64', 'int64'] and target_df[col].isnull().any():
            if 'mean' in col or 'rate' in col:
                target_df[col] = target_df[col].fillna(global_mean)
            elif 'median' in col:
                target_df[col] = target_df[col].fillna(target_df[col].median())
            else:
                target_df[col] = target_df[col].fillna(0)
    
    return target_df


# Create basic features
print("\n🔧 Creating comprehensive features...")
train_fe = create_comprehensive_features(train)
test_fe = create_comprehensive_features(test)
print(f"   ✅ Basic features created")

# DEBUG: Check raw data
print(f"\n🔍 RAW DATA CHECK:")
print(f"   Original 'liked' values: {train['liked'].unique()[:10]}")
print(f"   Original 'commented' values: {train['commented'].unique()[:10]}")
print(f"   Original 'clicked' values: {train['clicked'].unique()}")
print(f"   After processing 'clicked': {train_fe['clicked'].unique()}")

# Stratified K-Fold
n_folds = 5
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

# Initialize arrays
model_names = ['lgb1', 'xgb1', 'cat1', 'lgb2', 'xgb2', 'rf', 'et']
oof_preds = {name: np.zeros(len(train_fe)) for name in model_names}
test_preds = {name: np.zeros(len(test_fe)) for name in model_names}
cv_scores = {name: [] for name in model_names}

global_mean = train_fe['clicked'].mean()

print(f"\n{'='*70}")
print(f"🎯 Starting {n_folds}-Fold CV with {len(model_names)} Base Models")
print(f"{'='*70}\n")

for fold, (train_idx, val_idx) in enumerate(skf.split(train_fe, train_fe['clicked']), 1):
    print(f"\n{'─'*70}")
    print(f"📁 FOLD {fold}/{n_folds}")
    print(f"{'─'*70}")
    
    # Split
    train_fold = train_fe.iloc[train_idx].copy()
    val_fold = train_fe.iloc[val_idx].copy()
    test_fold = test_fe.copy()
    
    # Add aggregations - CRITICAL: Only use train_fold for stats!
    # DO NOT include val_fold when computing aggregations
    train_fold_agg = add_aggregation_features(train_fold, train_fold.copy(), global_mean)
    val_fold_agg = add_aggregation_features(train_fold, val_fold.copy(), global_mean)  # Use train_fold stats
    test_fold_agg = add_aggregation_features(train_fold, test_fold.copy(), global_mean)  # Use train_fold stats
    
    # Get features
    exclude = ['clicked', 'id', 'user_id', 'video_id', 'category', 'device', 
               'timestamp', 'watch_time_of_day']
    feature_cols = [c for c in train_fold_agg.columns if c not in exclude 
                    and train_fold_agg[c].dtype in ['int64', 'int32', 'float64', 'float32', 'int8', 'int16', 'uint8', 'bool']]
    
    X_train = train_fold_agg[feature_cols].replace([np.inf, -np.inf], 0).fillna(0)
    y_train = train_fold_agg['clicked'].values  # Use .values to ensure it's an array
    X_val = val_fold_agg[feature_cols].replace([np.inf, -np.inf], 0).fillna(0)
    y_val = val_fold_agg['clicked'].values  # Use .values to ensure it's an array
    X_test = test_fold_agg[feature_cols].replace([np.inf, -np.inf], 0).fillna(0)
    
    print(f"   Shape: Train={X_train.shape}, Val={X_val.shape}, Features={len(feature_cols)}")
    
    # CRITICAL DEBUG CHECK
    print(f"   🔍 Data sanity check:")
    print(f"      Target rate - Train: {y_train.mean():.3f}, Val: {y_val.mean():.3f}")
    print(f"      Target unique values: {np.unique(y_train)}")
    print(f"      Feature ranges OK: {X_train.isnull().sum().sum()==0 and X_val.isnull().sum().sum()==0}")
    
    # Check feature correlation with target
    if fold == 1:
        correlations = []
        for col in feature_cols[:20]:  # Check first 20 features
            corr, _ = spearmanr(X_train[col], y_train)
            correlations.append((col, corr))
        print(f"      Feature correlations with target:")
        for col, corr in sorted(correlations, key=lambda x: abs(x[1]), reverse=True)[:10]:
            print(f"         {col:40s}: {corr:+.4f}")
        
        # Check if aggregation features exist and have variance
        agg_features = [c for c in feature_cols if 'user_' in c or 'video_' in c or 'cat_' in c]
        if agg_features:
            print(f"      Aggregation features sample (first 5 rows):")
            print(X_train[agg_features[:5]].head())
    
    # ===== MODELS WITH GPU SUPPORT WHERE AVAILABLE =====
    models_config = []
    
    # LightGBM 1 - Try GPU with proper settings
    try:
        lgb1_gpu = lgb.LGBMClassifier(
            n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
            min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
            device='gpu', gpu_platform_id=0, gpu_device_id=0,
            gpu_use_dp=False,  # Use single precision (faster, more stable)
            random_state=42, verbose=-1)
        # Test if GPU works on small subset
        lgb1_gpu.fit(X_train[:1000], y_train[:1000])
        models_config.append(('lgb1', lgb.LGBMClassifier(
            n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
            min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
            device='gpu', gpu_platform_id=0, gpu_device_id=0,
            gpu_use_dp=False,
            random_state=42, verbose=-1)))
        if fold == 1: print("   [GPU] LightGBM 1")
    except Exception as e:
        models_config.append(('lgb1', lgb.LGBMClassifier(
            n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
            min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
            random_state=42, verbose=-1)))
        if fold == 1: print(f"   [CPU] LightGBM 1 (GPU failed: {str(e)[:50]})")
    
    # XGBoost 1 - Try GPU
    if xgb_gpu:
        models_config.append(('xgb1', XGBClassifier(
            n_estimators=600, learning_rate=0.05, max_depth=6,
            subsample=0.8, colsample_bytree=0.8, 
            tree_method='gpu_hist', gpu_id=0,
            random_state=42, eval_metric='logloss')))
        if fold == 1: print("   [GPU] XGBoost 1")
    else:
        models_config.append(('xgb1', XGBClassifier(
            n_estimators=600, learning_rate=0.05, max_depth=6,
            subsample=0.8, colsample_bytree=0.8, 
            tree_method='hist',
            random_state=42, eval_metric='logloss')))
        if fold == 1: print("   [CPU] XGBoost 1")
    
    # CatBoost 1 - Try GPU
    try:
        models_config.append(('cat1', CatBoostClassifier(
            iterations=500, learning_rate=0.05, depth=6,
            task_type='GPU', devices='0',
            verbose=0, random_state=42)))
        if fold == 1: print("   [GPU] CatBoost 1")
    except:
        models_config.append(('cat1', CatBoostClassifier(
            iterations=500, learning_rate=0.05, depth=6,
            verbose=0, random_state=42)))
        if fold == 1: print("   [CPU] CatBoost 1")
    
    # LightGBM 2 - Try GPU with proper settings
    try:
        lgb2_gpu = lgb.LGBMClassifier(
            n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
            min_child_samples=50, subsample=0.85, colsample_bytree=0.85,
            device='gpu', gpu_platform_id=0, gpu_device_id=0,
            gpu_use_dp=False,
            random_state=123, verbose=-1)
        # Test if GPU works
        lgb2_gpu.fit(X_train[:1000], y_train[:1000])
        models_config.append(('lgb2', lgb.LGBMClassifier(
            n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
            min_child_samples=50, subsample=0.85, colsample_bytree=0.85,
            device='gpu', gpu_platform_id=0, gpu_device_id=0,
            gpu_use_dp=False,
            random_state=123, verbose=-1)))
        if fold == 1: print("   [GPU] LightGBM 2")
    except Exception as e:
        models_config.append(('lgb2', lgb.LGBMClassifier(
            n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
            min_child_samples=50, subsample=0.85, colsample_bytree=0.85,
            random_state=123, verbose=-1)))
        if fold == 1: print(f"   [CPU] LightGBM 2 (GPU failed: {str(e)[:50]})")
    
    # XGBoost 2 - Try GPU
    if xgb_gpu:
        models_config.append(('xgb2', XGBClassifier(
            n_estimators=800, learning_rate=0.03, max_depth=5,
            subsample=0.8, colsample_bytree=0.8,
            tree_method='gpu_hist', gpu_id=0,
            random_state=999, eval_metric='logloss')))
        if fold == 1: print("   [GPU] XGBoost 2")
    else:
        models_config.append(('xgb2', XGBClassifier(
            n_estimators=800, learning_rate=0.03, max_depth=5,
            subsample=0.8, colsample_bytree=0.8,
            tree_method='hist',
            random_state=999, eval_metric='logloss')))
        if fold == 1: print("   [CPU] XGBoost 2")
    
    # RandomForest and ExtraTrees - CPU only
    models_config.append(('rf', RandomForestClassifier(
        n_estimators=300, max_depth=15, min_samples_split=10,
        max_features='sqrt', n_jobs=-1, random_state=42)))
    models_config.append(('et', ExtraTreesClassifier(
        n_estimators=300, max_depth=15, min_samples_split=10,
        max_features='sqrt', n_jobs=-1, random_state=42)))
    
    if fold == 1:
        print("   [CPU] RandomForest")
        print("   [CPU] ExtraTrees")
    
    for name, model in models_config:
        try:
            model.fit(X_train, y_train)
            val_pred = model.predict_proba(X_val)[:, 1]
            test_pred = model.predict_proba(X_test)[:, 1]
        except Exception as e:
            # If GPU fails, retry with CPU version
            if 'lgb' in name.lower() and 'gpu' in str(e).lower() or 'Check failed' in str(e):
                print(f"   {name.upper()} GPU failed, retrying with CPU...")
                if name == 'lgb1':
                    model = lgb.LGBMClassifier(
                        n_estimators=500, learning_rate=0.05, max_depth=6, num_leaves=31,
                        min_child_samples=50, subsample=0.8, colsample_bytree=0.8,
                        random_state=42, verbose=-1)
                elif name == 'lgb2':
                    model = lgb.LGBMClassifier(
                        n_estimators=700, learning_rate=0.03, max_depth=7, num_leaves=40,
                        min_child_samples=50, subsample=0.85, colsample_bytree=0.85,
                        random_state=123, verbose=-1)
                model.fit(X_train, y_train)
                val_pred = model.predict_proba(X_val)[:, 1]
                test_pred = model.predict_proba(X_test)[:, 1]
            else:
                raise e
        
        oof_preds[name][val_idx] = val_pred
        test_preds[name] += test_pred / n_folds
        score = roc_auc_score(y_val, val_pred)
        cv_scores[name].append(score)
        print(f"   {name.upper():6s} AUC: {score:.5f}")

# Print base model results
print(f"\n{'='*70}")
print("📊 BASE MODELS - CROSS-VALIDATION RESULTS:")
print(f"{'='*70}")
for name in model_names:
    oof_score = roc_auc_score(train_fe['clicked'], oof_preds[name])
    mean_cv = np.mean(cv_scores[name])
    std_cv = np.std(cv_scores[name])
    print(f"   {name.upper():6s} | OOF: {oof_score:.5f} | CV: {mean_cv:.5f} (±{std_cv:.4f})")

# ===== NEURAL NETWORK META-STACKER =====
print(f"\n{'='*70}")
print("🧠 TRAINING NEURAL NETWORK META-STACKER")
print(f"{'='*70}\n")

# Prepare meta features
meta_train = np.column_stack([oof_preds[name] for name in model_names])
meta_test = np.column_stack([test_preds[name] for name in model_names])

# Add extra features
extra_features = ['user_clicked_mean', 'video_clicked_mean', 'cat_clicked_mean', 
                  'watch_percent', 'engagement_score']
available_extra = [f for f in extra_features if f in train_fe.columns]

if available_extra:
    meta_train_extra = np.zeros((len(train_fe), len(available_extra)))
    for fold, (train_idx, val_idx) in enumerate(skf.split(train_fe, train_fe['clicked'])):
        train_fold = train_fe.iloc[train_idx].copy()
        val_fold = train_fe.iloc[val_idx].copy()
        val_fold_agg = add_aggregation_features(train_fold, val_fold, global_mean)
        for i, feat in enumerate(available_extra):
            if feat in val_fold_agg.columns:
                meta_train_extra[val_idx, i] = val_fold_agg[feat].fillna(0)
    
    meta_train = np.hstack([meta_train, meta_train_extra])
    test_agg = add_aggregation_features(train_fe, test_fe.copy(), global_mean)
    meta_test_extra = test_agg[available_extra].fillna(0).values
    meta_test = np.hstack([meta_test, meta_test_extra])

# Scale for neural network
scaler = StandardScaler()
meta_train_scaled = scaler.fit_transform(meta_train)
meta_test_scaled = scaler.transform(meta_test)

# Train neural networks with CV
nn_oof = np.zeros(len(train_fe))
nn_test = np.zeros(len(test_fe))

for fold, (train_idx, val_idx) in enumerate(skf.split(meta_train_scaled, train_fe['clicked']), 1):
    X_tr, X_val = meta_train_scaled[train_idx], meta_train_scaled[val_idx]
    y_tr, y_val = train_fe['clicked'].iloc[train_idx], train_fe['clicked'].iloc[val_idx]
    
    nn = MLPClassifier(hidden_layer_sizes=(128, 64, 32), activation='relu', 
                       max_iter=500, random_state=42, early_stopping=True)
    nn.fit(X_tr, y_tr)
    nn_oof[val_idx] = nn.predict_proba(X_val)[:, 1]
    nn_test += nn.predict_proba(meta_test_scaled)[:, 1] / n_folds
    
    score = roc_auc_score(y_val, nn_oof[val_idx])
    print(f"   Neural Net Fold {fold} AUC: {score:.5f}")

nn_oof_score = roc_auc_score(train_fe['clicked'], nn_oof)
print(f"\n   🧠 Neural Net OOF AUC: {nn_oof_score:.5f}")

# ===== FINAL ENSEMBLES =====
print(f"\n{'='*70}")
print("🏆 CREATING FINAL ENSEMBLE SUBMISSIONS")
print(f"{'='*70}\n")

test_stack = np.column_stack([test_preds[name] for name in model_names])

ensembles = {
    'sub_neural_net.csv': nn_test,
    'sub_weighted.csv': np.average(test_stack, weights=[0.18, 0.18, 0.18, 0.15, 0.11, 0.10, 0.10], axis=1),
    'sub_average.csv': test_stack.mean(axis=1),
    'sub_neural_blend.csv': (nn_test * 0.6 + test_stack.mean(axis=1) * 0.4),
    'sub_best_single.csv': test_preds[max(model_names, key=lambda x: roc_auc_score(train_fe['clicked'], oof_preds[x]))]
}

for name, pred in ensembles.items():
    binary_pred = (pred > 0.5).astype(int)
    pd.DataFrame({'id': test_fe['id'], 'clicked': binary_pred}).to_csv(name, index=False)
    print(f"   ✅ {name:25s} | Click rate: {binary_pred.mean():.3f}")

print(f"\n{'='*70}")
print("🎯 SUBMISSION PRIORITY:")
print("   1. sub_neural_net.csv       <- Neural network meta-learner")
print("   2. sub_neural_blend.csv     <- 60% NN + 40% ensemble")
print("   3. sub_weighted.csv         <- Weighted ensemble")
print("   4. sub_average.csv          <- Simple average")
print("   5. sub_best_single.csv      <- Best single model")
print(f"{'='*70}\n")

print(f"✨ Expected Leaderboard Score: ~{nn_oof_score:.3f} (±0.01-0.02)")
print(f"🎉 DONE! GPU acceleration enabled where available!")
```

```python
import pandas as pd
import numpy as np
import lightgbm as lgb
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score
from scipy.stats import spearmanr
import warnings
warnings.filterwarnings('ignore')

# ===== CUDA CHECK =====
import torch
print("="*70)
print("🔍 CUDA AVAILABILITY CHECK")
print("="*70)
print(f"PyTorch CUDA: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"Device: {torch.cuda.get_device_name(0)}")
xgb_gpu = torch.cuda.is_available()
print("="*70 + "\n")

print("="*70)
print("🚀 ULTIMATE ML PIPELINE - WITH NEURAL NETWORK")
print("="*70)

# Load data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

print(f"\n📊 Data Loaded:")
print(f"   Train: {train.shape}, Test: {test.shape}")
print(f"   Click rate: {train['clicked'].mean():.3f}")

# Remove watch_percent if exists (potential leakage)
if 'watch_percent' in train.columns:
    print(f"   ⚠️  Removing 'watch_percent' from original data")
    train = train.drop('watch_percent', axis=1)
    test = test.drop('watch_percent', axis=1) if 'watch_percent' in test.columns else test


def create_features(df):
    """Create all features."""
    df = df.copy()
    
    # Fix data types
    if df['liked'].dtype == 'object':
        df['liked'] = df['liked'].str.lower().replace({'yes': 1, 'no': 0})
    df['liked'] = pd.to_numeric(df['liked'], errors='coerce').fillna(0).astype(int)
    df['commented'] = pd.to_numeric(df['commented'], errors='coerce').fillna(0).astype(int)
    df['subscribed_after'] = pd.to_numeric(df['subscribed_after'], errors='coerce').fillna(0).astype(int)
    df['recommended'] = pd.to_numeric(df['recommended'], errors='coerce').fillna(0).astype(int)
    
    # Fix clicked
    if 'clicked' in df.columns:
        df['clicked'] = pd.to_numeric(df['clicked'], errors='coerce').fillna(0).astype(int).clip(0, 1)
    
    # Basic features
    df['watch_percent'] = (df['watch_time'] / (df['video_duration'] + 1)).fillna(0).clip(0, 5)
    df['engagement_score'] = df['liked'] + df['commented']*2 + df['subscribed_after']*3
    df['watch_complete'] = np.minimum(df['watch_percent'], 1.0)
    df['full_watch'] = (df['watch_percent'] >= 0.95).astype(int)
    df['quick_exit'] = (df['watch_percent'] < 0.1).astype(int)
    
    # Time features
    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    df['hour'] = df['timestamp'].dt.hour
    df['day_of_week'] = df['timestamp'].dt.dayofweek
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_prime_time'] = df['hour'].isin([18,19,20,21,22]).astype(int)
    
    # Device features
    df['is_mobile'] = (df['device'] == 'Mobile').astype(int)
    df['is_desktop'] = (df['device'] == 'Desktop').astype(int)
    
    # Category features
    for cat in ['Gaming', 'Music', 'Education', 'Sports']:
        df[f'is_{cat.lower()}'] = (df['category'] == cat).astype(int)
    
    return df


def add_aggregations(train_fold, target_df, global_mean):
    """Add aggregation features."""
    
    # User stats
    user_stats = train_fold.groupby('user_id').agg({
        'clicked': ['mean', 'count'],
        'watch_time': 'mean',
        'engagement_score': 'mean',
        'watch_percent': 'mean'
    })
    user_stats.columns = ['_'.join(col) for col in user_stats.columns]
    user_stats = user_stats.add_prefix('user_')
    
    # Video stats
    video_stats = train_fold.groupby('video_id').agg({
        'clicked': ['mean', 'count'],
        'watch_percent': 'mean'
    })
    video_stats.columns = ['_'.join(col) for col in video_stats.columns]
    video_stats = video_stats.add_prefix('video_')
    
    # Category stats
    cat_stats = train_fold.groupby('category')['clicked'].mean().to_frame('cat_click_rate')
    
    # Merge
    target_df = target_df.merge(user_stats, left_on='user_id', right_index=True, how='left')
    target_df = target_df.merge(video_stats, left_on='video_id', right_index=True, how='left')
    target_df = target_df.merge(cat_stats, left_on='category', right_index=True, how='left')
    
    # Fill missing
    for col in target_df.columns:
        if target_df[col].dtype in ['float64', 'int64'] and target_df[col].isnull().any():
            target_df[col] = target_df[col].fillna(global_mean if 'mean' in col or 'rate' in col else 0)
    
    return target_df


# Create features
print("\n🔧 Creating features...")
train_fe = create_features(train)
test_fe = create_features(test)
print(f"   ✅ Features created")

# Cross-validation
n_folds = 5
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

model_names = ['lgb1', 'xgb1', 'cat1']
oof_preds = {name: np.zeros(len(train_fe)) for name in model_names}
test_preds = {name: np.zeros(len(test_fe)) for name in model_names}
cv_scores = {name: [] for name in model_names}

global_mean = train_fe['clicked'].mean()

print(f"\n{'='*70}")
print(f"🎯 Starting {n_folds}-Fold CV")
print(f"{'='*70}\n")

for fold, (train_idx, val_idx) in enumerate(skf.split(train_fe, train_fe['clicked']), 1):
    print(f"\n📁 FOLD {fold}/{n_folds}")
    
    # Split
    train_fold = train_fe.iloc[train_idx].copy()
    val_fold = train_fe.iloc[val_idx].copy()
    
    # Add aggregations
    train_agg = add_aggregations(train_fold, train_fold.copy(), global_mean)
    val_agg = add_aggregations(train_fold, val_fold.copy(), global_mean)
    test_agg = add_aggregations(train_fold, test_fe.copy(), global_mean)
    
    # Get features
    exclude = ['clicked', 'id', 'user_id', 'video_id', 'category', 'device', 'timestamp', 'watch_time_of_day']
    feature_cols = [c for c in train_agg.columns if c not in exclude 
                    and train_agg[c].dtype in ['float64', 'int64', 'int32', 'float32', 'uint8', 'int8', 'bool']]
    
    X_train = train_agg[feature_cols].replace([np.inf, -np.inf], 0).fillna(0)
    y_train = train_agg['clicked'].values
    X_val = val_agg[feature_cols].replace([np.inf, -np.inf], 0).fillna(0)
    y_val = val_agg['clicked'].values
    X_test = test_agg[feature_cols].replace([np.inf, -np.inf], 0).fillna(0)
    
    print(f"   Features: {len(feature_cols)}, Train: {X_train.shape}, Val: {X_val.shape}")
    
    # Debug first fold
    if fold == 1:
        print(f"   Target: Train={y_train.mean():.3f}, Val={y_val.mean():.3f}")
        correlations = []
        for col in feature_cols[:15]:
            try:
                corr, _ = spearmanr(X_train[col], y_train)
                correlations.append((col, corr))
            except:
                pass
        print(f"   Top correlations:")
        for col, corr in sorted(correlations, key=lambda x: abs(x[1]), reverse=True)[:5]:
            print(f"      {col:30s}: {corr:+.4f}")
    
    # Models
    models = [
        ('lgb1', lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, max_depth=6, 
                                    random_state=42, verbose=-1)),
        ('xgb1', XGBClassifier(n_estimators=500, learning_rate=0.05, max_depth=6,
                               tree_method='gpu_hist' if xgb_gpu else 'hist',
                               random_state=42, eval_metric='logloss')),
        ('cat1', CatBoostClassifier(iterations=500, learning_rate=0.05, depth=6,
                                    task_type='GPU' if xgb_gpu else 'CPU',
                                    verbose=0, random_state=42))
    ]
    
    for name, model in models:
        try:
            model.fit(X_train, y_train)
            val_pred = model.predict_proba(X_val)[:, 1]
            test_pred = model.predict_proba(X_test)[:, 1]
        except Exception as e:
            if 'lgb' in name.lower():
                print(f"   {name.upper()} GPU failed, using CPU")
                model = lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, max_depth=6,
                                          random_state=42, verbose=-1)
                model.fit(X_train, y_train)
                val_pred = model.predict_proba(X_val)[:, 1]
                test_pred = model.predict_proba(X_test)[:, 1]
            else:
                raise e
        
        oof_preds[name][val_idx] = val_pred
        test_preds[name] += test_pred / n_folds
        score = roc_auc_score(y_val, val_pred)
        cv_scores[name].append(score)
        print(f"   {name.upper():6s} AUC: {score:.5f}")

# Results
print(f"\n{'='*70}")
print("📊 BASE MODELS - CROSS-VALIDATION RESULTS:")
print(f"{'='*70}")
for name in model_names:
    oof_score = roc_auc_score(train_fe['clicked'], oof_preds[name])
    mean_cv = np.mean(cv_scores[name])
    std_cv = np.std(cv_scores[name])
    print(f"   {name.upper():6s} | OOF: {oof_score:.5f} | CV: {mean_cv:.5f} (±{std_cv:.4f})")

# ===== NEURAL NETWORK META-STACKER =====
print(f"\n{'='*70}")
print("🧠 TRAINING NEURAL NETWORK META-STACKER")
print(f"{'='*70}\n")

# Prepare meta features (OOF predictions from base models)
meta_train = np.column_stack([oof_preds[name] for name in model_names])
meta_test = np.column_stack([test_preds[name] for name in model_names])

# Add some original features to meta
extra_features = ['user_clicked_mean', 'video_clicked_mean', 'cat_click_rate', 'watch_percent']
available_extra = [f for f in extra_features if f in train_fe.columns]

if available_extra:
    # For train: need to compute with CV to avoid leakage
    meta_train_extra = np.zeros((len(train_fe), len(available_extra)))
    for fold_idx, (train_idx, val_idx) in enumerate(skf.split(train_fe, train_fe['clicked'])):
        train_fold = train_fe.iloc[train_idx].copy()
        val_fold = train_fe.iloc[val_idx].copy()
        val_agg = add_aggregations(train_fold, val_fold.copy(), global_mean)
        for i, feat in enumerate(available_extra):
            if feat in val_agg.columns:
                meta_train_extra[val_idx, i] = val_agg[feat].fillna(0)
    
    meta_train = np.hstack([meta_train, meta_train_extra])
    
    # For test: use full train
    test_agg_full = add_aggregations(train_fe, test_fe.copy(), global_mean)
    meta_test_extra = test_agg_full[available_extra].fillna(0).values
    meta_test = np.hstack([meta_test, meta_test_extra])

# Scale for neural network
scaler = StandardScaler()
meta_train_scaled = scaler.fit_transform(meta_train)
meta_test_scaled = scaler.transform(meta_test)

# Train neural network with CV
nn_oof = np.zeros(len(train_fe))
nn_test = np.zeros(len(test_fe))

for fold_idx, (train_idx, val_idx) in enumerate(skf.split(meta_train_scaled, train_fe['clicked']), 1):
    X_tr, X_val = meta_train_scaled[train_idx], meta_train_scaled[val_idx]
    y_tr, y_val = train_fe['clicked'].iloc[train_idx].values, train_fe['clicked'].iloc[val_idx].values
    
    # Neural network: 3 hidden layers
    nn = MLPClassifier(hidden_layer_sizes=(128, 64, 32), activation='relu',
                       max_iter=500, random_state=42, early_stopping=True,
                       validation_fraction=0.1, n_iter_no_change=20)
    nn.fit(X_tr, y_tr)
    nn_oof[val_idx] = nn.predict_proba(X_val)[:, 1]
    nn_test += nn.predict_proba(meta_test_scaled)[:, 1] / n_folds
    
    score = roc_auc_score(y_val, nn_oof[val_idx])
    print(f"   Neural Net Fold {fold_idx} AUC: {score:.5f}")

nn_oof_score = roc_auc_score(train_fe['clicked'], nn_oof)
print(f"\n   🧠 Neural Net OOF AUC: {nn_oof_score:.5f}")

# Create submissions
print(f"\n{'='*70}")
print("🏆 CREATING SUBMISSIONS")
print(f"{'='*70}\n")

test_stack = np.column_stack([test_preds[name] for name in model_names])

ensembles = {
    'submission_neural_net.csv': nn_test,
    'submission_neural_blend.csv': (nn_test * 0.6 + test_stack.mean(axis=1) * 0.4),
    'submission_average.csv': test_stack.mean(axis=1),
    'submission_weighted.csv': np.average(test_stack, weights=[0.35, 0.35, 0.30], axis=1),
    'submission_best.csv': test_preds[max(model_names, key=lambda x: roc_auc_score(train_fe['clicked'], oof_preds[x]))]
}

for name, pred in ensembles.items():
    binary_pred = (pred > 0.5).astype(int)
    pd.DataFrame({'id': test_fe['id'], 'clicked': binary_pred}).to_csv(name, index=False)
    print(f"   ✅ {name:35s} | Click rate: {binary_pred.mean():.3f}")

print(f"\n{'='*70}")
print("🎯 SUBMISSION PRIORITY:")
print("   1. submission_neural_net.csv      <- Neural network meta-learner 🧠")
print("   2. submission_neural_blend.csv    <- 60% NN + 40% base models")
print("   3. submission_weighted.csv        <- Weighted ensemble")
print("   4. submission_average.csv         <- Simple average")
print(f"{'='*70}\n")

print(f"✨ Expected Score: ~{nn_oof_score:.3f}")
print(f"🎉 DONE! Neural network trained successfully!")
```
