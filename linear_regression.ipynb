```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
data = pd.read_csv('co2.csv')
```

```python
categorical_columns=['Make', 'Model', 'Transmission', 'Fuel Type', 'Vehicle Class']
encoder=OneHotEncoder(sparse_output=False, drop='first')
encoded_features=encoder.fit_transform(data[categorical_columns])
X=data.drop(columns=[ 'Make', 'Model', 'Transmission', 'Fuel Type', 'Vehicle Class']).values
X= np.hstack((X,encoded_features))
y=data['CO2 Emissions(g/km)'].values.reshape(-1, 1)


# Used sklearn only for the one hott encoding
```

```python
np.random.seed(42)
indices=np.arange(X.shape[0])
np.random.shuffle(indices)
split_index=int(0.8*len(indices))
train_indices=indices[:split_index]
test_indices=indices[split_index:]
X_train=X[train_indices]
y_train=y[train_indices]
X_test=X[test_indices]
y_test=y[test_indices]
X_train_bias=np.c_[np.ones((X_train.shape[0], 1)), X_train]
X_test_bias=np.c_[np.ones((X_test.shape[0], 1)), X_test]
```

```python
weights=np.random.randn(X_train_bias.shape[1],1) *0.01
learning_rate=0.001
iterations=1000
```

```python
for i in range(iterations):
    gradients=2/X_train_bias.shape[0]*X_train_bias.T@(X_train_bias@weights-y_train)
    weights=weights-learning_rate*gradients
    y_pred_train=X_train_bias@weights
    y_pred_test=X_test_bias@weights
```

```python
def calculate_r2(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2) 
    ss_residual = np.sum((y_true - y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)
    return r2
```

```python
test_mse=np.mean((y_pred_test-y_test)**2)
test_r2=calculate_r2(y_test,y_pred_test)
print(f"MSE-{test_mse}")
print(f"R2-{test_r2}")
```
