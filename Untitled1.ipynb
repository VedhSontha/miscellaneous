```python
import cv2
import numpy as np
from tensorflow.keras.models import model_from_json
from ultralytics import YOLO
import os

# --- Paths ---
video_path = r"C:\Users\vedhr\CODES\Obstacle avoidance\Fpv11.mp4"
output_path = r"C:\Users\vedhr\CODES\Obstacle avoidance\Fpv_hybrid_output.mp4"

# DroNet paths
model_json_path = r"C:\Users\vedhr\CODES\Obstacle avoidance\of-obstacledetection\DroNeTello\models\DroNet\model_struct.json"
model_weights_path = r"C:\Users\vedhr\CODES\Obstacle avoidance\of-obstacledetection\DroNeTello\models\DroNet\model_weights_new_best.h5"

# YOLO path (you'll train this)
yolo_weights_path = r"C:\Users\vedhr\CODES\Obstacle avoidance\semantic_drone_training\yolov8n_semantic\weights\best.pt"

# --- Load Models ---
print("Loading DroNet...")
with open(model_json_path, "r") as f:
    model_json = f.read()
dronet = model_from_json(model_json)
dronet.load_weights(model_weights_path)
print("✅ DroNet loaded!")

# Load YOLO (only if weights exist)
yolo_available = os.path.exists(yolo_weights_path)
if yolo_available:
    print("Loading YOLO...")
    yolo = YOLO(yolo_weights_path)
    print("✅ YOLO loaded!")
else:
    print("⚠️ YOLO weights not found. Running DroNet only mode.")
    print(f"Expected path: {yolo_weights_path}")
    yolo = None

# --- Preprocessing for DroNet ---
def preprocess_dronet(frame):
    """Preprocess frame for DroNet (200x200 grayscale)"""
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(gray, (200, 200))
    normalized = resized / 255.0
    return np.expand_dims(normalized, axis=(0, -1))  # (1, 200, 200, 1)

# --- Safety Assessment from YOLO Detections ---
def calculate_safety_from_yolo(detections, frame_width, frame_height):
    """
    Calculate crash probability based on obstacle positions and sizes
    
    Logic:
    - Obstacles in center of frame = higher danger
    - Obstacles in bottom half = closer = higher danger
    - Larger obstacles = higher danger
    """
    if detections is None or len(detections.boxes) == 0:
        return 0.0, "SAFE", []
    
    # Define danger zone
    # X: center 50% of frame (25% to 75%)
    # Y: bottom 60% of frame (40% to 100%)
    danger_x_min = frame_width * 0.25
    danger_x_max = frame_width * 0.75
    danger_y_min = frame_height * 0.4
    
    danger_score = 0.0
    dangerous_objects = []
    
    for box in detections.boxes:
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
        cls = int(box.cls[0].cpu().numpy())
        conf = float(box.conf[0].cpu().numpy())
        
        # Center of bounding box
        cx = (x1 + x2) / 2
        cy = (y1 + y2) / 2
        
        # Bounding box dimensions
        bbox_width = x2 - x1
        bbox_height = y2 - y1
        bbox_area = bbox_width * bbox_height
        
        # Check if in danger zone
        in_danger_zone = (danger_x_min < cx < danger_x_max) and (cy > danger_y_min)
        
        if in_danger_zone:
            # Calculate danger contribution
            
            # 1. Proximity factor (0-1, higher = closer/more dangerous)
            # Objects lower in frame are considered closer
            proximity = (cy - danger_y_min) / (frame_height - danger_y_min)
            
            # 2. Size factor (0-1, larger = more dangerous)
            size_factor = bbox_area / (frame_width * frame_height)
            
            # 3. Centrality factor (0-1, centered = more dangerous)
            center_offset = abs(cx - frame_width / 2) / (frame_width / 2)
            centrality = 1.0 - center_offset
            
            # Combined danger (weighted)
            object_danger = (
                proximity * 0.5 +      # 50% weight on proximity
                size_factor * 30 +     # Amplify size impact
                centrality * 0.3       # 30% weight on centrality
            )
            
            danger_score += object_danger * conf  # Weight by confidence
            
            dangerous_objects.append({
                'bbox': (x1, y1, x2, y2),
                'class': cls,
                'danger': object_danger
            })
    
    # Clamp danger score to [0, 1]
    danger_score = min(danger_score, 1.0)
    
    # Determine status
    if danger_score > 0.5:
        status = "CRASH RISK!"
    elif danger_score > 0.25:
        status = "CAUTION"
    else:
        status = "SAFE"
    
    return danger_score, status, dangerous_objects

# --- Main Processing ---
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    raise Exception(f"Cannot open video: {video_path}")

# Video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Output writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

print(f"\nProcessing video: {video_path}")
print(f"Output will be saved to: {output_path}")
print("Press 'q' to stop.\n")

frame_count = 0
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    frame_count += 1
    
    # --- 1. DroNet Steering Prediction ---
    dronet_input = preprocess_dronet(frame)
    pred = dronet.predict(dronet_input, verbose=0)[0]
    steering_angle = float(pred[0])
    
    # --- 2. YOLO Obstacle Detection (if available) ---
    if yolo_available:
        detections = yolo(frame, verbose=False)[0]
        
        # Calculate crash probability from detections
        crash_prob, status, dangerous_objects = calculate_safety_from_yolo(
            detections, width, height
        )
        
        # Use YOLO's annotated frame
        annotated_frame = detections.plot()
        
        # Highlight dangerous objects in red
        for obj in dangerous_objects:
            x1, y1, x2, y2 = obj['bbox']
            cv2.rectangle(annotated_frame, 
                         (int(x1), int(y1)), (int(x2), int(y2)),
                         (0, 0, 255), 3)  # Red thick box
        
    else:
        # Fallback: use hacky heuristic from steering
        crash_prob = min(max(abs(steering_angle), 0.0), 1.0)
        status = "CRASH RISK!" if crash_prob > 0.5 else "SAFE"
        annotated_frame = frame.copy()
    
    # --- 3. Overlay Information ---
    # Background rectangles for better readability
    cv2.rectangle(annotated_frame, (5, 5), (400, 130), (0, 0, 0), -1)
    cv2.rectangle(annotated_frame, (5, 5), (400, 130), (255, 255, 255), 2)
    
    # Text overlays
    cv2.putText(annotated_frame, f"Steering: {steering_angle:.3f}", 
                (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    
    # Crash probability color-coded
    crash_color = (0, 255, 0) if crash_prob < 0.25 else (0, 165, 255) if crash_prob < 0.5 else (0, 0, 255)
    cv2.putText(annotated_frame, f"Crash Prob: {crash_prob:.3f}", 
                (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, crash_color, 2)
    
    # Status color-coded
    status_color = (0, 255, 0) if status == "SAFE" else (0, 165, 255) if status == "CAUTION" else (0, 0, 255)
    cv2.putText(annotated_frame, f"Status: {status}", 
                (10, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.7, status_color, 2)
    
    # Model source
    model_text = "DroNet + YOLO" if yolo_available else "DroNet only"
    cv2.putText(annotated_frame, model_text, 
                (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
    
    # --- 4. Draw Steering Indicator (Arrow) ---
    arrow_y = height - 60
    arrow_start = (width // 2, arrow_y)
    arrow_end = (int(width // 2 + steering_angle * 250), arrow_y)
    
    # Arrow background
    cv2.line(annotated_frame, (50, arrow_y), (width - 50, arrow_y), (100, 100, 100), 2)
    
    # Steering arrow
    arrow_color = (0, 255, 255)  # Yellow
    cv2.arrowedLine(annotated_frame, arrow_start, arrow_end, arrow_color, 5, tipLength=0.3)
    
    # Steering label
    cv2.putText(annotated_frame, "STEERING", (width // 2 - 60, arrow_y - 20),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
    
    # --- 5. Progress Bar ---
    progress = frame_count / total_frames
    bar_width = width - 20
    bar_height = 15
    bar_x = 10
    bar_y = height - 30
    
    cv2.rectangle(annotated_frame, (bar_x, bar_y), 
                 (bar_x + bar_width, bar_y + bar_height), (50, 50, 50), -1)
    cv2.rectangle(annotated_frame, (bar_x, bar_y), 
                 (bar_x + int(bar_width * progress), bar_y + bar_height), 
                 (0, 255, 0), -1)
    cv2.putText(annotated_frame, f"{frame_count}/{total_frames}", 
                (bar_x + 5, bar_y + 12), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (255, 255, 255), 1)
    
    # --- 6. Write and Display ---
    out.write(annotated_frame)
    
    cv2.imshow('Hybrid Obstacle Avoidance', annotated_frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("\nStopped by user.")
        break
    
    # Print progress every 30 frames
    if frame_count % 30 == 0:
        print(f"Progress: {frame_count}/{total_frames} frames ({progress*100:.1f}%)")

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"\n{'='*60}")
print("✅ PROCESSING COMPLETE!")
print(f"{'='*60}")
print(f"📹 Output saved to: {output_path}")
print(f"📊 Processed {frame_count} frames")
print(f"🤖 Models used: {'DroNet + YOLO' if yolo_available else 'DroNet only'}")
```
